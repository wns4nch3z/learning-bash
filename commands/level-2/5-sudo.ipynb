# SUDO (superuser do)
`sudo` permite ejecutar comandos específicos como otro usuario (por defecto root).

> Permiso temporal

## Sintaxis
```bash
sudo comando
sudo -u usuario comando
```

> Pide tu propia contraseña, no la de root.

## ¿Cómo funciona internamente?
  * Verifica `/etc/sudoers`
  * Eleva privilegios **solo para el comando**
  * Registra quién se ejecutó qué
  * Tiempo limitado (timestamp)

## INGRESO ROOT
### `sudo su`
Funciona, pero no es buena práctica

### `sudo su -`
Mejor versión

### `sudo -i`
Es la forma moderna y recomendada
