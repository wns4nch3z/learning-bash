# GROUPADD
Crear grupos en Linux

## Sintaxis
```bash
groupadd [opciones] grupo
```
> Los grupos sirven para **gestionar permisos colectivos**

````python
!sudo groupadd red 
````

````python
!tail -n 3 /etc/group
````

## Opciones
### `-g` 
GID específico

Útil para:
  * Servidores
  * Sincronizar usuarios
  * NFS / LDAP


````python
!sudo groupadd -g 2000 developers
````

````python
!tail -n 3  /etc/group
````

### `-r`
Grupo del sistema

Crea GID bajo (ej. <1000)

````python
!sudo groupadd -r backups
````

````python
!tail -n 3 /etc/group
````

### `-f` 
Forzar creación
  * No da error
  * No hace nada

````python
!sudo groupadd -f developers # Ya esta creado
````

````python
!tail -n 3 /etc/group
````

Otra forma de verificar:

````python
!getent group developers
````
