# CHMOD | Change mode
Sirve para modificar los permisos de lectura (r), escritura (w) y ejecución (x) de un **archivo o carpeta**

**Revisar:**
  * [Permisos](../../notebooks/10-permisos.ipynb)
  * [Asignaciones | Simbólico y octal](../../notebooks/11-asignaciones.ipynb)

## Sintaxis
```bash
chmod [opciones] modo archivo
```
  * `modo` puede ser simbólico o numérico
  * `archivo` es el nombre del archivo o carpeta



```bash
# Simbólico > Agrega + | Quita -
chmod u+x archivo.sh 
chmod g+x archivo.sh
chmod o-w archivo.sh

# Numérico
chmod 755 archivo.sh
```

## Opciones
### `-R`
Aplicar cambios recursivamente a directorios y subdirectorios

### `--verbose`
Mostrar cambios realizados

### `--reference=archivo_ref`
Copiar permisos de otro archivo
