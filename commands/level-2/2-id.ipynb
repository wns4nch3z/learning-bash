## Sintaxis
```bash
id [OPTION]... [USER]...
```

````python
# Uso simple
!id
````

Muestra:
  * **UID**: User ID -> en este caso usuario normal
  * **GID**: Group ID -> grupo principal del usuario
  * **Groups**: Grupos secundarios

````python
# Uso con otro usaurio
!id root 
````

> Nota: Si estás en sudo, puedes escalar a root fácilmente.

## Opciones

### `-u`
Solo UID

````python
!id -u
````

### `-g`
Solo GID

````python
!id -g
````

### `-G`
Solo grupos

````python
!id -G
````

### `-Gn`
Grupo con nombres

````python
!id -Gn
````
