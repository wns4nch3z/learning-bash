# SU (SWITCH USER)
`su` permite cambiar de usuario, normalmente a `root`


## Sintaxis
```bash
su [opciones] [usuario]
```

> Si no indicas usuario, por defecto es `root`.


```bash
# Uso simple
su
# Cambia root 
su root
# Otro usuario
su otroUsuario
```

> Pide la contraseña del usuario destino (root o el usuario al que cambias).

## Opciones
### `-l` (login)
```bash
su -l usuario
```

> Igual que `su -`

### `-c` (ejecutar comando)
```bash
su -c "comando" usuario
```

### `-s` (shell)
```bash
# Usar una shell específica
su -s /bin/bash usuario
```






## Notas
### `su`
  * No login completo
  * Entorno heredado (mal)

### `su -`
  * Login completo
  * Forma correcta
