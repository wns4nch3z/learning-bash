| Grupo    | Riesgo                 |
| -------- | ---------------------- |

| `wheel`  | Root en muchas distros |
| `docker` | Root indirecto         |
| `lxd`    | Root indirecto         |
| `adm`    | Logs sensibles         |
| `shadow` | Hashes de passwords    |
