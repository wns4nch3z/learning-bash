# USERDEL
Elminar usuarios en Linux

## Sintaxis
```bash
userdel [opciones] usuario
```
Elimina el usuario del sistema
  * No siempre elimina el HOME
  * No elimina archivos fuera del HOME
  * No elimina grupos automáticamente

## Opciones

### `-r` 
Elimina HOME y mail

Borra: `/home/juan`, `/var/mail/juan`

```bash
userdel -r juan
```

### `-f`
Forzar eliminación
```bash
userdel -f juan
```
Elimina aunque: 
  * Usuario esté logueado
  * HOME no exista

## Nota
Los archivos NO se borran automáticamente si están fuera del HOME.

Buscar archivos
```bash
find / -user juan 2>/dev/null
```

Después de borrar:
```bash
ls -l archivo.txt
# dueño: UID 1002
```
> Quedan archivos huérfanos
