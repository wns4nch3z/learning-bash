# USERADD
Crear usuarios en Linux

## Sintaxis
```bash
useradd [opciones] usuario
```
  * No crea contraseña por defecto
  * No se crea HOME si no se indica (depende de distro)






````python
# Ver usuarios
!tail -n 3 /etc/passwd
````

````python
!sudo useradd will
````

````python
# Ver añadido
!tail -n 3 /etc/passwd
````

````python
!ls /home/ # No hay HOME de will
````

> Añadido pero con shell por defecto

## Opciones

### `-m` 
Crear directorio HOME

````python
!sudo useradd -m juan
````

````python
# Ver añadido
!tail -n 3 /etc/passwd
````

> Shell por defecto

````python
# Ver HOME creado
!ls /home/
````

### `-d` 
Definir un HOME personalizado

````python
!sudo useradd -m -d /opt/luis luis
````

````python
# Home elegido
!ls /opt
````

### `-s` 
Definir una shell

````python
!sudo useradd -s /bin/bash maria
````

````python
# Ver añadido
!tail -n 3 /etc/passwd
````

### `-G`
Grupos secundarios

````python
!sudo useradd -m -G red,developers jose
````

````python
!tail -n 4 /etc/group
````

### `-g`
Grupo primario

````python
!sudo useradd -m -g developers pedro
````

````python
!tail -n 7 /etc/passwd
````

> GUID es de Developers

### `-u`
UID específico

````python
!sudo useradd -u 1500 rosa
````

````python
!tail -n 3 /etc/passwd
````

### `-c` 
Comentario / descripción

````python
!sudo useradd -c "Usuario de pruebas" lara
````

````python
!tail -n 3 /etc/passwd
````

### `-e` 
Fecha de expiración

`useradd -e 2026-01-01 ricardo`

### `-f`
Días antes de desactivar (password expira)

`useradd -f 7 juan`

## Ver los grupos de un usuario

````python
# Ejemplos
!groups jose
````

## Crear contraseña (obligatorio)
`passwd nombre_usuario`

## Ejemplo completo
```bash
sudo useradd -m -s /bin/bash -G sudo -c "Usuario admin" juan
passwd juan
# Asignar contraseña
```
  * `-m`: HOME creado
  * `-s`: SHELL personalizada
  * `-G`: Grupo secundario sudo
  * `-c`: Agregar comentario
