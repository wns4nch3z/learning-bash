# GROUPDEL
Eliminar grupos en Linux

  * No elimina usuarios
  * No elimina grupos 
  * Solo borra el grupo

## Ejemplo simple
```bash
sudo groupdel testgroup
```
El grupo desaparece de:
  * `/etc/group`
  * `/etc/gshadow`

Verificación:
  * `getent group testgroup`


## Pasa después

### Usuarios
Verificar con `groups nombre_usuario`

> Grupo se elimina silenciosamente

### Archivos
Los archivos no se eliminan, pero quedan con un GID huérfano
```bash
# Ejemplo
ls -l archivo.txt
-rw-r----- 1 juan testgroup archivo.txt # Antes
-rw-r----- 1 juan 1003 archivo.txt # Después
```
> El grupo ya no existe, solo queda el número.




## FORMA CORRECTA
### 1. Ver si tiene usuarios
`getent group testgroup`
### 2. Quitar usuarios del grupo
`gpasswd -d juan testgroup`
### 3. Buscar archivos del grupo
`find / -group testgroup 2>/deb/null`
### 4. Cambiar grupo de archivos
`chgrp users archivo.txt`
### 5. Eliminar grupo
`groupdel testgroup`
