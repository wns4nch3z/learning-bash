# CHOWN
Cambiar dueño y grupo

## ¿Cómo funciona?
```bash
chown propietario:grupo archivo
```

Cambia: Propietario (user) y grupo (group)

## Sintaxis
```bash
chown [opciones] usuario archivo # Solo propietario
chown [opciones] usuario:grupo archivo # Propietario y grupo
chown [opciones] :grupo archivo # Solo grupo
```

## Opciones

### `-R` 
Cambiar recursivamente
```bash
chown -R juan:developers /var/www
```
> Cambia todo dentro del directorio

### `-v`
Modo verbose

### `--reference`
Copiar permisos de otro archivo
```bash
chown --reference=archivo1 archivo2
```
