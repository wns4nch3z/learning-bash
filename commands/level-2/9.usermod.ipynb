# USERMOD
Modificar usuarios existentes

Si useradd crea -> usermod ajusta, afina, repara


## SINTAXIS
```bash
usermod [opciones] usuario
```
  * NO crea usuario
  * NO crea HOME (solo lo mueve si le dices)


## Opciones

### 1. `-aG`
Agregar grupos secundarios

````python
!sudo usermod -aG developers juan
````

````python
# AÃ±adido grupo dev
!groups juan
````

> Siempre usa `-aG` juntos, por que `-G` solo, borra otros grupos

### 2. `-g`
Cambiar grupo primario

````python
# Ejemplo grupos primario | GUID
!id jose 
````

````python

!sudo usermod -g developers jose
````

````python
!id jose # Cambio GUID
````

### `-s`
Cambiar shell

````python
# Ejmplo con Lara
!!tail -n 1 /etc/passwd
````

````python
!sudo usermod -s /bin/bash lara
````

````python
!tail -n 1 /etc/passwd # Cambio a bash
````

### `-d` 
Cambiar HOME

````python
!sudo usermod -d /opt/juan juan
````

### `-m`
Mover HOME (usar con -d)

````python
!sudo usermod -d /opt/pedro -m pedro
````

````python
!ls /opt # Movido
````

### `-u`
Cambiar UID

````python
!id juan
````

````python
!sudo usermod -u 2500 juan
````

````python
!id juan
````

### `-c`
Cambiar comentario

### `-l`
Cambiar nombre de usuario
