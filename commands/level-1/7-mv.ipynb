# MV 
El comando sirve para mover archivos/carpeta a otra ubicación o renombrarlos sin copiarlos.

## Sintaxis
```bash
mv [opciones] origen destino
```

### Renombrando un archivo

````python
!mv ../../files/borrador/carpeta-existente/test3/backup/texto4.txt ../../files/borrador/carpeta-existente/test3/renombrado.txt && ls ../../files/borrador/carpeta-existente/test3
````

### Moviendo archivos

````python
!mv ../../files/borrador/carpeta-existente/test3/backup/texto3.txt ../../files/borrador/carpeta-existente/test3/movido.txt && ls ../../files/borrador/carpeta-existente/test3
````

## Opciones

| Opción | Qué hace                                                  | Ejemplo                |
| ------ | --------------------------------------------------------- | ---------------------- |
| `-i`   | Pregunta antes de sobreescribir                           | `mv -i a.txt b.txt`    |
| `-f`   | *Force*: no pregunta, sobreescribe directo                | `mv -f a.txt b.txt`    |
| `-v`   | Muestra qué está moviendo                                 | `mv -v a.txt /tmp/`    |
| `-u`   | Solo mueve si el archivo destino es más viejo o no existe | `mv -u a.txt /backup/` |
