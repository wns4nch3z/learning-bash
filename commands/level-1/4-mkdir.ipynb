# MKDIR | Make Directory
Se usa para crear directorios en el sistema de archivos

## Sintaxis
```bash
mkdir [opciones] [Ruta]
```

````python
!ls -l ../../files/borrador/carpeta-existente/ # Lista de directorio actual
````

````python
!mkdir ../../files/borrador/carpeta-existente/test1 ../../files/borrador/carpeta-existente/test2 ../../files/borrador/carpeta-existente/test3
````

> Puede crear una o varias carpetas

````python
!ls -l ../../files/borrador/carpeta-existente/
````

## Opciones

### `-p`
Crear directorios padres si no existen
  * Si falta algún directorio intermedio, lo crea sin errores.
  * También evita error si el directorio ya existe.

````python
!mkdir -p ../../files/borrador/carpeta-existente/test1/padre/hijo/nieto
````

````python
!tree ../../files/borrador/carpeta-existente/
````

> Comando `tree` muestra directorios y archivos como un árbol

### `-v` | Modo verbose
Muestra lo que va haciendo

````python
!mkdir -v ../../files/borrador/carpeta-existente/test2/padre/
````

### `-m` | Establecer permisos
Permite crear el directorio con **permisos**(después) específicos
