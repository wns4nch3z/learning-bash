# RM | REMOVE
Sirve para eliminar archivos y directorios de manera permanente (no van a una papelera, se borran de verdad).

## Sintaxis
```bash
rm [opciones] []
```

````python
!rm ../../files/borrador/carpeta-existente/test2/texto?.txt && ls ../../files/borrador/carpeta-existente/test2/
````

## Opciones
| Opción | Qué hace                               | Ejemplo             |
| ------ | -------------------------------------- | ------------------- |
| `-r`   | Eliminar directorios recursivamente    | `rm -r backup/`     |
| `-f`   | Forzar: no pregunta ni muestra errores | `rm -f secreto.txt` |
| `-v`   | Mostrar qué está eliminando            | `rm -v *.log`       |
| `-i`   | Pregunta antes de borrar               | `rm -i archivo.txt` |
