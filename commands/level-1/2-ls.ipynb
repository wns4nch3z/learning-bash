````python
!ls
````

## Opciones
### `-l` | Formato largo
Muestra permisos, propietario, grupo, tamaño, fecha, nombre de archivo.


````python
!ls -l
````

### `-a` 
Mostrar archvivos ocultos 

> Se pueden combinar opciones

````python
!ls -la ../../files/borrador/
````

### `-h` 
Tamaños legibles (Lectura humana) de más a menos


````python
!ls -lh ../../files/test/ # 3.9k, 4.4k
````

### `R` | Listado recursivo
Muestra subdirectorios y su contenido

````python
!ls -R ../../files/borrador/
````

### `-t` | Ordenar fecha
Archivos más recientes primero

````python
!ls -lt ../../files/borrador
````

### `-S` | Ordenar por tamaño
Grandes -> pequeños


````python
!ls -lhS ../../files/test/
````

### `-d`
Muestra solo los nombres de directorios y evitar listar contenido.
  * `*/` → solo directorios dentro de esa carpeta

````python
!ls -d ../../*/
````
