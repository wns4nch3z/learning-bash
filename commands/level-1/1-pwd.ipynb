Muestra la ruta completa (Ruta absoluta) del directorio actual donde estás ubicado en la terminal.



## Sintaxis
```bash
pwd [opciones]
```

````python
!pwd
````

> Por defecto: sin opciones ya es útil.

## Opciones
| Opción | Descripción                                             | Uso típico                            |
| ------ | ------------------------------------------------------- | ------------------------------------- |
| `-P`   | Muestra la ruta **real**, sin seguir enlaces simbólicos | Cuando existe un enlace tipo shortcut |
| `-L`   | Muestra la ruta **lógica** (incluye symlinks)           | Por defecto en muchas distros         |
