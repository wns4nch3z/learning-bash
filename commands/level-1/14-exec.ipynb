En linux todo es un archivo (archivos, carpetas, procesos, dispositivos, etc.)

Cada programa recibe 3 canales básicos, numeros así:
| Número | Nombre | Función                   |
| ------ | ------ | ------------------------- |
| **0**  | stdin  | Entrada (lo que escribes) |
| **1**  | stdout | Salida normal             |
| **2**  | stderr | Errores                   |


## Función de `exec`
Sirve para:
  * Manipular los descriptores
  * Redirigir stdin, stdout y stderr permanentemente
  * Ejecutar programas reemplazando el shell actual



### `exec` usado para redirigir FD 

#### a. Rederigir stdout
```bash
exec 1>salida.txt
```

> Todo lo que escriba stdout irá al archivo

#### b. Rederigir stderr
```bash
exec 2>errores.txt
```
> La salida de errores se guarda en el archivo

#### c. Abrir un nuevo descriptor 
```bash
exec 3>log.txt
```
***Se usa así:***
```bash
echo "Guardando en FD3" >&3
```
> Esto no va a stdout, va a `log.txt`

***Cerrar un FD***
```bash
exec 3>&-
```

#### d. Leer desde un archivo
```bash
exec 4< entreda.txt
```
***Leer desde FD***
```bash
cat <&4
```

#### e. Copiar entre descriptores
```bash
exec 5>&4
```
> Tendra una copia del otro descriptor


#### f. Copiar y cerrar un descriptor
```bash
exec 6>&4-
```
> Copia y cierra el descriptor FD 4

#### g. Reemplaza el shell actual
```bash
exec comando
```
No crea un proceso hijo.
> Reemplaza el shell actual por ese comando.

```bash
exec python3
exec bash
```
