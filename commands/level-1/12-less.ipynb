# LESS 
Es un visor de texto interactivo. Muestra los archivos de manera paginada, desplazarte hacia arriba/abajo, buscar, saltar líneas, etc.

## Sintaxis
```bash
less [opciones] archivo
```

También puedes usarlo recibiendo datos por tubería: `comando | less`

Uso simple: `less archivo`
  * `enter`: Avanza línea por línea
  * `espacio`: Avanza por página
  * `b`: Sube una página 
  * `\buscartexto`: Escribir `/`, ingresar la palabra y navegar por los resultados `n` para próximo o `N` para el anterior.
  * `q`: Para salir del modo paginado 

## Opciones

### `-i` 
Si quieres que la búsqueda distinga entre mayúsculas y minúsculas

### `+F` ! Modo follow(seguimiento)
  * Similar `tail -f` 
  * Pausar follow `ctrl + c`
  * Puedo usar `/buscar`, `n`, `N`
  * Reanudar el seguimiento `F`
