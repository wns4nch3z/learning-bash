# TOUCH 
Es un comando que usa para:
  * Crear archivos vacíos si no existen.
  * Actualizar las fechas de acceso (atime) y modificación (mtime) de un archivo existente

> No modifica el contenido del archivo, solo su "timestamp" (fechas)


## Sintaxis
```bash
touch [opciones] archivo1 archivo2 ... # uno o varios archivos
```

> Si el archivo no existe, lo crea

````python
!touch ../../files/borrador/carpeta-existente/test1/vacio1.txt ../../files/borrador/carpeta-existente/test1/vacio2.txt && ls ../../files/borrador/carpeta-existente/test1 
````

````python
# PRACTICA
!ls -l ../../files/borrador/carpeta-existente/test1
````

````python
# Ver estado
!stat ../../files/borrador/carpeta-existente/test1/vacio1.txt
````

## Opciones

### `-a`
Actualizar solo fecha de acceso

````python
!touch -a ../../files/borrador/carpeta-existente/test1/vacio1.txt && stat ../../files/borrador/carpeta-existente/test1/vacio1.txt
````

### `-m`
Actualiza la solo la fecha de modificación

````python
# Modifica el access
!touch -m ../../files/borrador/carpeta-existente/test1/vacio1.txt && stat ../../files/borrador/carpeta-existente/test1/vacio1.txt
````

### `t`
Cambiar fecha/hora específica
**Formato:** [[CC]YY]MMDDHHMM[.ss]

````python
!stat ../../files/borrador/carpeta-existente/test1/vacio2.txt # Fecha original
````

````python
!touch -t 200101111111 ../../files/borrador/carpeta-existente/test1/vacio2.txt && stat ../../files/borrador/carpeta-existente/test1/vacio2.txt
````

### `r`
Usar la fecha de otro archivo

`touch -r origen.txt destino.txt`

> Origen archivo del cual se copia la fecha, destino es el que recibe la fecha

````python
!touch -r ../../files/test/prueba.txt  ../../files/borrador/carpeta-existente/test1/vacio2.txt && stat ../../files/borrador/carpeta-existente/test1/vacio2.txt
````
