# XARGS
Lee elementos de STDIN (línea por línea o separadps por espacios/newlines) y los convierte en argumentos para otro comando. Es el puente entre salida de un comando y argumentos de otro.

  * 1. STDIN -> flujo de texto (pipes `|`)
  * 2. Argumentos de línea de comandos -> lo que va después del comando

Existe porque muchos comandos no leen de stdin, solo leen argumentos como rm, cp, mv, touch, mkdir, chmod, chown, ls, kill.

> `xargs` convierte texto recibido por STDIN en argumentos de un comando.



## Sintaxis
```bash
comando_que_genera_texto | xargs comando_destino
```

````python
!echo ../../files/borrador/fake.png | xargs rm
````

> Termina ejecutando: `rm ../../files/borrador/fake.png`

## Opciones

### `-n`
Controlar argumentos se pasan

````python
!cat ../../files/test/prueba.txt | xargs -n 3 echo
````

### `-I`
Cuando necesitas uno por uno

````python
!cat ../../files/test/prueba.txt | head -n 4 | xargs -I {} echo "Número: {}"
````

> {} es un placeholder.

## Extra

### Delimitadores
### `-0`
En linux separa por:
  * espacios
  * tabs
  * nuevas líneas

> Usar NULL (\0) como separador, algo que nunca aparece en nombres de archivo por eso `-0` para trabajar con ellos.

```bash
comando | xargs -0 rm
```


Explicado después con `find` usando `-print0` 
```bash
# Ejemplo
find . -name "*.txt" -print0 | xargs -0 rm
```

### `-d`
Permite delimitadores

````python
!echo "a:b:c" | xargs -I {} -d ":" echo "Soy {}"
````
