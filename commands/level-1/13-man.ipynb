# MAN | Manual
Muestra el manual de un comando, con su descripción, opciones, ejemplos y más 


## Sintaxis
```bash
man comando
```

## Uso 

````python
!man ls
````

> `q`: Para salir y `space` avanzar línea por línea.

### Secciones

#### NAME
Explica que hace el comando
```bash
ls -list directory contents 
```

#### SYNOPSIS
Muestra la forma correcta de usar el comando
```bash
ls [OPTION]... [FILE]...
```

#### DESCRIPTION
Explica con detalle lo que hace el comando


#### OPTIONS
Lista todas las opciones que puedes usar, con explicación
```bash
-l    use a long listing format
-a    show hidden files
```

#### BUSCAR
Escribir `/` y después lo que se busca navegar con `n` siguiente coincidencia y `N` anterior coincidencia 

#### Interpretar synopsis
Símbolos:
  * `[]` (Opcional): No es obligatorio -> `ls [FILE]` (ls solo funciona)
  * `...` (tres puntos): Puede repetirse -> ls [OPTIONS]... [FILE]...
  *  `|` : Elige entre uno y otro
  * `Sin []`: Obligatorio -> `cp SOURCE DESTINATION`
  * `algo`: Variable que tú debes reemplazar, si ves algo como FILE, DIRECTORY, PATTERN significa que debes poner tú
