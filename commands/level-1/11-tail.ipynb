# TAIL | FINAL
Sirve para ver las últimas líneas de una archivo


## Sintaxis
```bash
tail [opciones] archivo
```

````python
!tail ../../files/test/prueba.txt  # Muestra las últimas 10 líneas
````

## Opciones

### `n`
Indicar cuántas líneas ver

````python
!tail -n 5 ../../files/test/prueba.txt
````

### `c`
Mostrar bytes en lugar líneas

````python
!tail -c 3 ../../files/test/prueba.txt
````

### Seguimiento
Es un modo donde muestra lo que se va agregando en tiempo real

````python
!tail -f ../../files/test/prueba.txt
````
