# TREE
Nos muestra la estructura de directorios y archivos en forma de árbol, lo que facilita visualizar jerarquías de carpetas.

## Sintaxis
```bash
tree [opciones] [ruta]
```
  * Ruta: directorio que quieres visualizar (por default usa el actual)

````python
!tree # Default
````

## OPCIONES
### `-d`
Solo directorios

````python
!tree -d ../../files
````

### `-L`
Limitar la profundidad

````python
# muestra solo 1 nivel de profundidad
!tree -L 1 ../../../bash-scripting/
````

````python
!tree -L 1 -d ../../files/
````

````python
!tree -a # Por defecto aparece al final la cantidad de archivos y directorios
````

### `-I` 
Ignorar archivos


````python
!tree -I "test.txt" ../../files/test/
````
