# ECHO 

`echo` se usa para imprimir texto o **variables** en la terminal.

## Sintaxis
```bash
echo [opciones] texto
```


````python
# Ejemplo simple
!echo "Hola"
````

## Opciones
### `-n` 
No imprime un salto de línea al final

> En notebooks se usa "!" antes de los comandos para ejecutarlos

````python
!echo "Hola"  # Hace un salto
!echo "Mundo"
!echo -n "Sin " #Evita el salto
!echo "Salto"
````

### `-e` 
Habilita interpretación de caracteres especiales (\) como:
  * `\n` -> Nueva línea
  * `\t` -> tabulación
  * `\\` -> Barra invertida
  * `\"` -> Comillas dobles 

````python
!echo -e "Nueva\nlínea"
!echo -e "Saludo \"Hola Mundo\""
!echo -e "Ruta windows C:\\Usuarios\\TuNombre\\Documentos\\mi_archivo.txt"
````
