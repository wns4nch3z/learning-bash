# HEAD | Comienzo
Muestra el comienzo de un archivo, sirve para ver las primeras líneas sin abrirlo completo.

## Sinxtaxis
```bash
head [opciones] archivo
```

````python
!head ../../files/test/prueba.txt # Default primeas 10 líneas
````

## Opciones

### `-n` 
Cambia el número predeterminado de filas a mostrar

````python
!head -n 5 ../../files/test/prueba.txt
````

### `-c` 
Cuenta en  bytes (o sea, caracteres)

1 carácter = 1 byte

````python
!head -c 10 ../../files/test/archivos-datos.txt
````

### Múltiples archivos

````python
!head -n 2 ../../files/test/prueba.txt ../../files/test/access.logs
````
