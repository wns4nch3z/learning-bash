Sirve para contar líneas, palabras, caracteres y bytes en archivos o en la entrada estándar (STDIN)

Puede mostrar:
  * Líneas
  * Palabras
  * Caracteres
  * Longitud de la línea más larga



## SINTAXIS
```bash
wc [OPCIONES] [ARCHIVO...]
```

````python
# USO BÁSICO
!wc ../../files/test/prueba.txt
````

> Muestra que hay 19 líneas, 20 palabras, 50 bytes

````python
# USO CON STDIN
!echo "Hola mundo" | wc
````

## Opciones


### `-l`
Muestra solo el número de líneas

````python
!wc -l ../../files/test/prueba.txt
````

````python
!cat -E ../../files/test/prueba.txt
````

> Cuenta saltos de línea (\n), no “líneas” visibles.

### `-w`
Muestra solo el número de palabras

````python
!wc -w ../../files/test/prueba.txt
````

### `-c`
Muestra solo el número de bytes

>Cuenta cuántos bytes reales ocupa el archivo.

````python
!wc -c ../../files/test/prueba.txt
````

````python
!echo "áéíóú" | wc -c
````

### `-m`
Muestra solo el número de caracteres

  * Cuenta caracteres Unicode (también llamados “code points”).
  * Un solo carácter puede ocupar 1, 2, 3 o 4 bytes en UTF-8.

````python
!wc -m ../../files/test/prueba.txt
````

````python
!echo "áéíóú" | wc -m # 6 caracteres cuenta el salto de línea de echo
````

### `-L`
Muestra la longitud de la línea más larga
