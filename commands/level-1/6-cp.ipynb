## Opciones


### `-r` | Recursivo
Copiar directorios completos

````python
!cp -r ../../files/borrador/carpeta-existente/test2/ ../../files/borrador/carpeta-existente/test3/backup && ls ../../files/borrador/carpeta-existente/test3/
````

### `-i`
Pregunta antes de sobreescribir  

````python
!cp -i ../../files/borrador/carpeta-existente/test2/texto4.txt ../../files/borrador/carpeta-existente/test3/backup/
````

### `-v` | Verbose
Muestra que se esta copiando

### `-u` 
Solo copia si el archivo destino no existe o es mÃ¡s viejo

### `-p`
Mantiene permisos, fechas y propietarios

### `-a`
Copia exacta del archivo/dir manteniendo todo (modo archivo)

> `-a` = `-r` + `-p` + copia de enlaces -> Ideal para respaldo de sistemas
