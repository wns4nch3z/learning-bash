# LOCATE | Búsqueda por índice
Busca archivos por nombre usando una base de datos indexada, no el disco en tiempo real.

> `locate` consulta una base de datos creada por `updatedb`



**Ventajas**
  * Instantáneo
  * No recorre el disco
  * Ideal para búsquedas rápidas

**Desventajas**
  * Puede mostrar archivos eliminados
  * No muestra archivos recién creados
  * Depende de cuándo se ejecutó updatedb

> Es rápido, pero no siempre exacto.

## SINTAXIS
```bash
locate [OPCIONES] PATRÓN
```

````python
# Uso simple
!locate texto1.txt
````

## OPCIONES
### `-i` 
Ignorar mayúsculas/minúsculas

````python
!locate -i "TEXTO1.txt"
````

### `-n` 
Limitar resultados
```bash
locate -n 3 archivo
```
> Sirve en kali

### `-c` 
Contar resultados

````python
!locate -c passwd
````

### `-r`
Expresiones regulares
  * `.` -> Cualquier carácter
  * `^` -> Inicio
  * `$` -> Fin
  * `*` -> 0 o más
  * `?` -> 0 o 1
  * `(abc)` -> Grupos y OR
  * `[abc]` -> Caracteres `a o b o c`
  * `[0-9]` -> números
  * `[a-z]` -> letras

> Para que tome de forma literal `()`, `*`, `.`, `/` se usa el `\` para que no lo tome como un regex.

````python
!locate -r "texto[1-4]\.txt$"
````

````python
!locate -r 'texto[13]\.txt'
````

### `--existing`
solo archivos que aún existen

````python
!locate --existing texto1.txt
````
