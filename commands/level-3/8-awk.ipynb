# AWK
Sirve para **procesar texto línea por línea**, especialmente útil para:
  * Archivo de texto 
  * Salida de comandos (`ls`, `ps`, `df`, etc.)
  * Archivo CSV, logs, reportes

## Sintaxis
```bash
awk 'patrón {acción}' archivo
```
O usando la salida de otro comando:
```bash
comando | awk 'patrón {acción}'
```
  * **patrón** -> condición (opcional)
  * **acción** -> lo que quieres hacer
  * Si no hay patrón -> 




````python
# Preparando
!echo -e "Juan 25 Lima\nAna 30 Bogotá\nPedro 22 Madrid" > ../../files/test/practica.txt
````

````python
# Leer 
!cat ../../files/test/practica.txt
````

````python
%%bash
# Otra forma de ejecutar en notebook con %%bash
awk '{print}' ../../files/test/practica.txt
````

> Imprime todas las líneas (igual que `cat`).

### Trabajando con columnas (campos)
`awk` divide cada línea por espacios (por defecto).

* `$1` → primera columna
* `$2` → segunda
* `$0` → toda la línea
* `$<Número de columna>` -> otra cualquier columna


````python
%%bash
awk ' {print $1}' ../../files/test/practica.txt
````

````python
%%bash
awk '{print $1, $3}' ../../files/test/practica.txt
````

### Usando condiciones (patrones)
Mostrar solo personas mayores de 25:

````python
%%bash
awk '$2 > 25 {print $1, $2}' ../../files/test/practica.txt
````

  * `$2 > 25` es el patrón
  * `{print $1, $2}` es la acción

### Separadores de campo
Cambiar el separador por defecto con `-F`

````python
%%bash
# Separador en este caso es :
awk -F: '{print $2}' ../../files/test/test.txt | head -n 4
````

### Variaciones internas útiles

| Variable | Significado        |
| -------- | ------------------ |
| `NR`     | Número de línea    |
| `NF`     | Número de columnas |
| `$NF`    | Última columna     |




````python
%%bash
awk -F: '{print NR, $2}' ../../files/test/test.txt
````

````python
%%bash
awk '{print $1, $NF}' ../../files/test/practica.txt
````

### Bloques BEGIN y END
  * `BEGIN` se ejecuta antes de leer el archivo
  * `END` se ejecuta al final

````python
%%bash
awk 'BEGIN {print "INICIO"} {print $1} END {print "FIN"}' ../../files/test/practica.txt
````

### Operaciones matemáticas

````python
%%bash
awk -F: '{suma += $5} END {print suma/NR}' ../../files/test/test.txt 
````

### Condiciones avanzadas
Operadores:
  * `==` igual  
  * `!=` distinto
  * `&&` AND
  * `||` OR


````python
%%bash
awk -F: '$5 >= 2000 && $5 <= 2500 {print $2}' ../../files/test/test.txt 
````

### Formato de salida
Usando `printf`

````python
%%bash
awk -F: '{printf "Nombre: %-15s | Saldo: %d\n", $2, $5}' ../../files/test/test.txt
````
