# TR
`tr` se usa para **traducir, reemplazar o eliminar caracteres** de la entrada estándar (**STDIN**) y enviar el resultado a la salida estándar (STDOUT).

No trabaja directamente con archivos: se usa normalmente con **pipes (`|`)** o redirecciones.

## Sintaxis básica
```bash
tr [opciones] SET1 [SET2]
```
  * **SET1**: conjunto de caracteres a buscar
  * **SET2**: conjunto de caracteres por los que se reemplazan
  * Si no hay `SET`, suele usarse con opciones como eliminar (`-d`)

Internamente, tr:
  * Lee byte por byte (no líneas)
  * Compara cada byte con SET1
  * Si hay match:
    * Sustituye
    * Elimina
    * Comprime (según opciones)
  * Escribe inmediatamente la salida

#### Datos
`tr` reemplaza por posición, no por nombre

Cuando haces: `tr SET1 SET2`
> “el carácter nº1 de SET1 → se convierte en el carácter nº1 de SET2
el carácter nº2 de SET1 → se convierte en el carácter nº2 de SET2
etc.”

Es una tabla de correspondencia por índice, no por significado.

Ejemplo mínimo: `echo "abc" | tr 'abc' 'XYZ'`

Internamente construye así
| Posición | SET1 | SET2 |
| -------- | ---- | ---- |
| 1        | a    | X    |
| 2        | b    | Y    |
| 3        | c    | Z    |

```ouput
XYZ
```
> Parece "normal"

**¿SET es más largo?**
```bash
echo "hola" | tr "hola" "adios" 
```
Se podría pensar:
  > “hola → adios”

Eso no pasa

Tabla real:
| Pos | SET1 | SET2 |
| --- | ---- | ---- |
| 1   | h    | a    |
| 2   | o    | d    |
| 3   | l    | i    |
| 4   | a    | o    |


```ouput
adio
```
> No es magia, es posicional.


**¿SET2 más corto?**
```bash
echo "abcabc" | | tr "abc" "x"
```
Regla importante:
  > Si `SET2` es más corto, se repite su último carácter

Tabla efectiva:
| SET1 | SET2 |
| ---- | ---- |
| a    | X    |
| b    | X    |
| c    | X    |

```bash
XXXXXX
```
Esto se usa mucho para normalizar cosas:
```bash
tr '[:digit:]' '0'
```
Todo número -> `0`

> `[:digit:]` es una clase caracteres POSIX, este representa dígitos decimales
