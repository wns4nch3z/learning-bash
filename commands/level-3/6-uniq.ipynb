# UNIQ
`uniq` detecta líneas repetidas consecutivas y las elimina o las cuenta.


**Dato:**
uniq solo funciona bien si las líneas iguales están juntas.
Por eso casi siempre se usa con sort.

## Sintaxis 
```bash
uniq [opciones] [archivo_entrada] [archivo_salida]
```

````python
!cut -d: -f6 ../../files/test/test.txt | sort | uniq
````

## Opciones

### `-c`
Contar repeticiones

````python
!cut -d: -f6 ../../files/test/test.txt | sort | uniq -c
````

### `-d`
Mostrar solo duplicados

````python
!cut -d: -f6 ../../files/test/test.txt | sort | uniq -d
````

### `-u`
Mostrar solo líneas únicas

````python
!cut -d: -f6 ../../files/test/test.txt | sort | uniq -u
````

### `-i`
Ignorar mayúsculas/minúsculas 
```bash
sort nombres.txt | uniq -i
```

### `-w`
Comparar solo N caracteres
```bash
uniq -w3 archivo.txt
```
