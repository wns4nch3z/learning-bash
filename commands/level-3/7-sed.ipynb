# SED
`sed` significa **Stream Editor** (editor de flujo)

`sed` **no trabaja por columnas**, trabaja **por líneas completas** y su principal uso es:
  * Buscar texto
  * Reemplazar texto
  * Elimina líneas
  * Modificar texto **en tiempo real** (pipes)

Muy usado en:
  * Logs
  * Configuraciones
  * Limpieza de texto
  * Automatizació

## Sintaxis
```bash
sed 'comando' archivo
```
O con pipe:
```bash
comando | sed 'comando'
```
Imprimir texto (comportamiento básico):

````python
!sed '' ../../files/test/test.txt | head -n 3
````

> Imprime todo (igual que `cat`).

## Usos
### `s` | substitute
Sustitución > Solo cambia **la primera coincidencia por línea**.


````python
!sed 's/Madrid/MADRID/' ../../files/test/test.txt | head -n 4
````

> Solo reemplaza **la primera coincidencia de cada línea**.

### `g` | global
Reemplazo global > Todas las coincidencias

````python
!sed "s/Sales/SALES/g" ../../files/test/test.txt |tail -n 5
````

### `i`
Ignorar mayúsculas/minúsculas

````python
!sed 's/MADRID/Limita/gi' ../../files/test/test.txt
````

### `-n`
Mostrar solo líneas modificadas
  * `-n` → no imprimas todo
  * `p` → imprime solo si hubo sustitución

````python
!sed -n '' ../../files/test/test.txt # Ejemplo 
````

````python
!sed -n 's/Madrid/LIMA/p' ../../files/test/test.txt
````

### `d`
Borrado de líneas

````python
!sed '1d' ../../files/test/test.txt | head -n 4 # Primera línea borrada
````

````python
!sed '1,4d' ../../files/test/test.txt | head -n 4 # Varias líneas
````

### Mostrar rango de líneas

````python
!sed -n '5,8p' ../../files/test/test.txt
````

### Reemplazar líneas
#### En líneas específica

````python
!sed '/IT/ s/Developer/DEV/' ../../files/test/test.txt # Solo IT que tenga Developer
````

#### Solo línea concreta

````python
!sed -n '3 s/Developer/DEV/p' ../../files/test/test.txt
````

### Modificar archivo realmente `-i`
```bash
sed -i 's/Madrid/MADRID/g' archivo.txt
```
