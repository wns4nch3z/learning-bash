# SORT
`sort` lee líneas de texto (de archivos o stdin) y las reordena según criterios definidos. Por defecto ordena lexicográficamente (como diccionario), línea por línea.

**Por qué existe**: Ordenar datos es una operación fundamental en procesamiento de texto, análisis de logs, reportes, y pipelines de datos Unix.

## Sintaxis
```bash
sort [opciones] [archivo]
```
También puede usarse con pipes:
```bash
comando | sort [opciones]
```

````python
!cut -d: -f2 ../../files/test/test.txt | sort
````

Sin opciones, `sort` ordena: 
  * Línea completa como unidad
  * Alfabéticamente (ASCII)
  * Ascendente (A→Z)
  * Sensible a mayúsculas (mayúsculas antes que minúsculas en ASCII)

Ejemplo de uso:
```bash
cat frutas.txt
# Manzana
# banana
# Arándano
# cereza

sort frutas.txt
# Arándano
# Manzana
# banana    ← minúscula va después
# cereza
```

> Importante: El ordenamiento ASCII es: números < Mayúsculas < minúsculas

````python
# Ejemplo
!echo -e "8\n2\nA\n4\nz\na" | sort
````

## Opciones
### `n`
Orden numérico 

````python
!cut -d: -f5 ../../files/test/test.txt | sort -n
````

### `-r` 
Reversa (descendente)

````python
!sort -nr ../../files/test/prueba.txt
````

### `-u`
Muestra elementos únicos (elimina duplicados)

````python
!cut -d: -f6 ../../files/test/test.txt | sort -u
````

> Equivalente a sort | uniq pero más eficiente.

### `-k` 
Ordenar por columna / campo específico

Esto es crítico para datos tabulares. Define qué parte de cada línea usar para ordenar.

Sintaxis: `-k INICIO[,FIN]`

```bash
# Datos separados por espacios (delimitador por defecto)
cat empleados.txt
# Juan 35 Ventas
# Ana 28 IT
# Pedro 42 Ventas
# Laura 28 Marketing

# Ordenar por edad (columna 2)
sort -k2 -n empleados.txt
# Ana 28 IT
# Laura 28 Marketing
# Juan 35 Ventas
# Pedro 42 Ventas

# Ordenar por departamento (columna 3), luego por edad
sort -k3,3 -k2,2n empleados.txt
# Ana 28 IT
# Laura 28 Marketing
# Juan 35 Ventas
# Pedro 42 Ventas
```

### `-t`
Delimitador de campos.

Por defecto, `sort` usa espacios/tabs. Con `-t` defines otro:
 
```bash
# CSV con comas
cat datos.csv
# nombre,edad,ciudad
# Juan,35,Madrid
# Ana,28,Barcelona
# Pedro,42,Madrid

# Ordenar por ciudad (campo 3)
sort -t',' -k3 datos.csv
# Ana,28,Barcelona
# Juan,35,Madrid
# Pedro,42,Madrid
```

> Cuidado: -t acepta un solo carácter.

### `-h` 
Ordenamiento "humano" de tamaños

```bash
# Tamaños de archivos
du -h /var/log/* 2>/dev/null | sort -h
# 4.0K    /var/log/boot.log
# 128K    /var/log/messages
# 2.3M    /var/log/syslog
# 45M     /var/log/journal
```
> Entiende sufijos: K, M, G, T, P.

### `-v` 
Ordenamiento de versiones
```bash
echo -e "file-1.10.txt\nfile-1.2.txt\nfile-1.9.txt" | sort
# file-1.10.txt   ← lexicográfico incorrecto
# file-1.2.txt
# file-1.9.txt

echo -e "file-1.10.txt\nfile-1.2.txt\nfile-1.9.txt" | sort -V
# file-1.2.txt    ← orden correcto de versiones
# file-1.9.txt
# file-1.10.txt
```

## NOTA
Lexicográfico = comparar como strings de texto, carácter por carácter, usando códigos ASCII.

  * ✓ Perfecto para texto (nombres, palabras, códigos)
  * ✗ Incorrecto para números puros (usa -n)
  * ✗ Incorrecto para versiones (usa -V)
  * ✗ Incorrecto para tamaños human-readable (usa -h)
