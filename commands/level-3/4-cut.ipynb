# CUT
`cut` extrae columnas de texto a partir de:
  * **delimitadores** (`:`, `,`, `|` etc)
  * **posiciones fijas** (caracteres)

Es ideal para archivos tipo:
  * `/etc/passwd`
  * logs
  * CSV simples
  * salidas de comandos

> `cut` **no entiende datos**, **no evalúa condiciones**, **no formatea**. Solo **corta** texto según **posiciones** o **delimitadores**.

## Estructura interna de un archivo 
Tu archivo es **texto plano** con un delimitador fijo `:`.

Ejemplo (una línea):

```
1:Juan Perez:IT:SysAdmin:2500:Madrid:2018-03-15
```

Internamente, `cut` lo ve así:

| Campo | Contenido    |
| ----- | ------------ |
| 1     | `1`          |
| 2     | `Juan Perez` |
| 3     | `IT`         |
| 4     | `SysAdmin`   |
| 5     | `2500`       |
| 6     | `Madrid`     |
| 7     | `2018-03-15` |



## Sintaxis básica

```bash
cut -d DELIMITADOR -f CAMPOS archivo
```

| Parte     | Qué hace            |
| --------- | ------------------- |
| `-d`      | define el separador |
| `-f`      | campos que quieres  |
| `archivo` | entrada             |

---

````python
# Uso simple: Muestre nombres
# cut -d ":" -f "2,4" archivo o cut -d: -f2,4 archivo
!cut -d: -f2 ../../files/test/test.txt | head -n 3
````

### Qué pasa por dentro:

1. Lee línea por línea
2. Busca `:`
3. Divide la línea en campos
4. Devuelve **solo el campo 2**

````python
#Extraer varios campos: Nombre y ciudad
!cut -d: -f2,6 ../../files/test/test.txt | head -n 3
````

````python
!cut -d ":" -f "2,4" ../../files/test/test.txt | head -n 3
````

* `-c` → caracteres
* `--complement` → todo MENOS lo indicado

## Opcioens
### `cut` por caracteres `-c`
Cuando NO hay delimitador.

````python
!echo "ABCDEFG" | cut -c 2,4,6
````

### `--complement` (poco conocido)
Todo menos lo indicado

````python
!cut -d: -f2 --complement ../../files/test/test.txt | head -n 3
````
