El comando `ln` se utiliza para crear enlaces duros o enlaces simbólicos.

## SINTAXIS
```bash
ln [opciones] origen destino
```


## Opciones

### Crer un enlace duro | HARD LINK
Ambos nombres apuntan al mismo inodo.

`ln` por defecto sin opciones crea enlaces duros.


````python
!mkdir -p /tmp/test/enlace
````

````python
!ls /tmp/test/
````

````python
!echo "Hola" > /tmp/test/original.txt
````

````python
!ls -li /tmp/test && cat /tmp/test/original.txt
````

> Con ls -li nos muestra el inode en la primera columna y también el contador de enlaces `1` (default)

````python
# Creando enlace duro 
!ln /tmp/test/original.txt /tmp/test/enlace/duro.txt && ls -li /tmp/test/enlace/duro.txt
````

> Se ve que muestra en el mismo inode y el contador es 2 ahora

  * No hay original y copia
  * Ambos nombres so niguaales
  * Si editas uno modifica ya que es el mismo archivo

````python
!cat /tmp/test/original.txt # Viendo el original
!echo "Adiós" >> /tmp/test/enlace/duro.txt # Editando el archivo duro
!cat /tmp/test/original.txt # Modificacion
````

**Borrando el archivo**

````python
!rm /tmp/test/original.txt
````

````python
!ls -li /tmp/test/enlace/ && ls -li /tmp/test/
````

¿Se borra el archivo? -> NO

  * Un solo inodo con múltiples nombres
  * El archivo vive mientras exista al menos un nombre

### Crear un enlace simbólico | SYMLIK
Para crear un archivo simbólico usamos la opción `-s`.

````python
# Usaremos el duro.txt para crear un enlace simbólico
!ln -s /tmp/test/enlace/duro.txt /tmp/test/simbólico.txt && ls -li /tmp/test/simbólico.txt
````

Se nos muestra enlace de quién es. Además el tipo de archivo es representado por una `l`

  * Inodo es distinto
  * Contenido: texto de la ruta

#### Diferencias de borrado
| Acción                    | Enlace duro       | Enlace simbólico   |
| ------------------------- | ----------------- | ------------------ |
| Borras el nombre original | El archivo sigue  | El enlace se rompe |
| Borras el enlace          | Nada pasa         | Nada pasa          |
| Borras el destino         | Sigue funcionando | Se rompe           |
