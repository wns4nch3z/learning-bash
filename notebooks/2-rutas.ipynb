Para movernos en Linux necesitamos comprender dor formas de indicar rutas:
  * Ruta absoluta
  * Ruta relativa

## Rutas
### 1. Ruta absoluta | Absolute Path
> Empieza desde la raíz `/`

Una ruta absoluta siempre comienza con `/` (la raíz del sistema)

Es el camino completo **desde la raíz hasta un archivo o carpeta**.

Ejemplo:
  * /etc/passwd
  * /home/usuario


```python
!pwd # Muestra una ruta absoluta
```

```python
!ls /home/ 
```

### 2. Ruta relativa | Relative Path
> Se basa en tu posición actual

Una ruta relativa no empieza con `/`.

Se interpreta a partir del directorio donde te encuentras en ese momento.

Ejemplo: 
  * Documents/
  * etc/archivo.txt

```python
!ls # Toma mi ruta actual
```

## Carateres especiales

### `~` | Directorio actual
Representa el directorio personal del usuario actual es decir `/home/usuarioactual`

```python
# Ejemplo de usuarios
!ls /home
```

```python
# Manual
!ls /home/user/
```

```python
# Caracter ~
!ls ~
```

### `.` | Directorio actual
El directorio donde estás parado ahora mismo.

```python
!ls . # . representa directorio actual
```

### `..` | Directorio padre
El directorio superior al actual

```python
!ls ..
```

```python
!ls ../../../ # Aplicando varios
```
