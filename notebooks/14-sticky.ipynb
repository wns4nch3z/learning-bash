# STICKY BIT
El **Sticky Bit** se utiliza principalmente en **directorios**. Su propósito es evitar que los usuarios borren o renombren archivos que no les pertenecen en directorios compartidos.

```bash
# Ejemplo es /tmp
drwxrwxrwt 2 root root 4096 dic 12 14:30 /tmp
```

> El sticky es representado por una `t` en lugar de `x`

Si ves que un directorio como `/tmp` tiene el **Sticky Bit** activado, aunque muchos usuarios puedan escribir en él, no podrán borrar o modificar los archivos de otros usuarios. 

## Asignación
### Simbólico
Con `sudo chmod +t directorio`

### Numérico
Representado por 1 (1000)

Con `sudo chmod 1777 directorio`


## Notas
La `T` mayúscula representa que no esta activada la "x"
