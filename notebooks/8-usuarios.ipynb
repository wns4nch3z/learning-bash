# Usuarios
En linux todo gira alrededor de usuarios y permisos
  * Usuario normal
  * Usaurio root (superusuario) 

## 1. Usuario normal 
Es el usuario limitado, el que normalmente obtienes al explotar una vulnerabilidad.

### Características
  * UID >= 1000 (normalmente)
  * No puede modificar el sistema
  * Solo puede acceder a sus propios archivos
  * Necesita permisos especiales para tareas administrativas

### Como identificar
Con el comando `whoami`:
```bash
user_actual
```
Con el comando `id`:
```bash
uid=1001(user) gid=1001(user) groups=1001(user)
```

## 2. Usuario root (superusuario)
Es el máximo privilegio del sistema

### Características
  * UID=0
  * Control total
  * Puede leer, escribir y ejecutar todo
  * Cambiar permisos
  * Persistencia

### Como identificar
Con el comando `whoami`:
```bash
root
```
Con el comando `id`:
```bash
uid=0(root) gid=0(root) groups=0(root)
```

## 3. Ver usuarios del sistema
### Archivos clave `/etc/passwd`
Archivo que contiene la información básica de los usuarios (nombres, ID de usuario, directorio home, shell, etc).

Cada línea representa cada usuario y cada campo está separado por dos puntos `:`.

`cat /etc/passwd`
```bash
# Campos
nombre_de_usuario:contraseña:UID:GID:comentario:directorio_home:shell
```
  * Usuario: Nombre del usuario dentro del sistema
  * Contraseña: Originalmente usado para almacenar la contraseña cifrada del usuario. Por ahora por seguridad suele contener `x` o `*` para indicar que la contraseña esta cifrada en `/etc/shadow`
  * UID: User ID -> >= 1000 o 0 (root)
  * GID: Group ID -> Cada usuario pertenece a un grupo
  * Comentario: Información adicional
  * Directorio home: Directorio personal del usuario -> `/home/user`
  * Shell: Interprete de comandos predeterminado para el usuario.

> Cualquier usuario con UID 0 es root, aunque no se llame “root”.

```bash
# Campos
juan:x:1001:1001:Juan Pérez:/home/juan:/bin/bash
```

#### Usuarios de servicios
`www-data`, `mysql`, `ftp` ...
  * No suelen tener shell
  * Si son comprometidos, pueden llevarte a root

```bash
www-data:x:33:33:www-data:/var/www:/usr/sbin/nologin
```
