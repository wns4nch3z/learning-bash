# PROCESOS
Un proceso es:

> un programa que está ejecutándose

Ejemplo:
  * El archivo `ls` -> programa
  * Cuando ejecutas `ls` -> proceso

Cada proceso tiene:
  * PID -> identificador único
  * Usuario que lo ejecuta
  * Estado
  * Recursos (CPU, memoria)

## PID y PPID
  * PID: Process ID
  * PPID: PID del proceso padre

```bash
echo $$
```
Muestra el PID de tu shell actual

Ver árbol de procesos:
```bash
pstree
```

## Ver procesos en ejecución
Ver todos los procesos: `ps aux`

Columnas importantes:
  * USER -> quién lo ejecuta
  * PID
  * %CPU
  * %MEM
  * COMMAND

Ver procesos en tiempo real
```bash
top
```
> O `htop` si lo tienes
