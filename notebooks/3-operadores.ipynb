# Operadores de control de shell


### `&&` | Ejecutar solo si el anterior tuvo éxito
**Función:** Ejecuta el comando de la derecha solo si el comando de la izquierda terminó correctamente. 


```python
!pwd && ls -l
```

> Si algún comando falla, los siguientes no se ejecutan.

### `||` | Ejecutar solo si el anterior falló
**Función:** Ejecuta el comando de la derecha solo si el comando de la izquierda falla (exit code ≠ 0).

```python
!ls ../files/borrador/carpeta-inexistente/ || echo "La carpeta no existe"
```

### `;` | Ejecutar siempre
**Función**: Ejecuta el siguiente comando siempre, sin importar si el anterior falló o tuvo éxito.

```python
!ls ../files/borrador/carpeta-inexistente; echo "La carpeta no existe"; pwd
```

> Se pueden combinar estos operadores
