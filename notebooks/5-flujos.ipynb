# Flujos Estándar 


## STDIN | Entrada Estándar -> 0
stdin (file descriptor 0) es el canal de entrada de un programa: es dónde el programa lee datos.

En una terminal normal:
  * Tu teclado -> se conecta por stdin
  * Lo que escribes -> entra por stdin
  * El programa lee desde stind como si fuera un archivo.

> Nota: No es lo que el programa muestra, es lo que el programa recibe para procesar.

### STDIN como entrada implícita
Hay comandos que automáticamente "adivinan" stdin cuando no les das argumentos:
  * `cat` (muestra stdin)
  * `grep` (filta stdin)
  * `sort` (ordena stdin)
  * `wc` (cuenta stdin)

### Ejemplo
```bash
!cat #Queda esperando datos, porque está leyendo desde stdin
# Escribes y va stdin
# cat lee stdin -> y lo envía a stdout (por eso aparece en pantalla)
# Para terminar el flujo stdin: ctrl + d (EOF)
```

## EOF | End of File
Es una señal lógica, NO un carácter visible(`\n` o una tecla del teclado).

Indica que no hay más datos para leer en stdin.

## STDOUT | Salida estándar -> 1
  * Es el canal donde un programa envía su salida normal (no errores).
  * Es el flujo que se conecta a tuberías y redirecciones (después)


### Ejemplos

````python
# Ejemplo 1
!echo "Hola Mundo" # La respuesta que ves proviene de stdout
````

````python
# Ejemplo 2
!ls
````

````python
# Ejemplo 3
!cat ../files/test/prueba.txt # El contenido del archivo es stdout
````

## STDOUT | Error estándar -> 2
stderr (standard error) es el flujo de errores estándar de un programa.

Es el canal donde un programa escribe:
  * mensajes de error
  * advertencias
  * información crítica


### ¿Cuándo usa un programa stderr?
#### Errores (los más comunes)
  * archivo no encontrado
  * permiso denegado
  * opción inválida
  * comando no existe
#### Avisos (warnings)
Programas avanzados pueden enviar advertencias a stderr

#### Diagnósticos
Si un programa imprime logs o información técnica, es común que use stderr para no contaminar stdout

### Ejemplos

````python
# Ejemplo 1 | Archivo inexiste
!ls noexiste # La salida es stderr
````

````python
# Ejemplo 2 | Comando mal escrito
!abcd # stderr
````

````python
# Ejemplo 3 | Opción inváida
!ls --opcioninventada #stderr
````
