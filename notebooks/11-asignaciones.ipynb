# Asignación de permisos
Hay dos maneras de agregar permisos simbólico y octal.

## 1. Simbólico 
En este modo a las clases de usuarios se les asignan una letra para representarlos y a los permisos `rwx` se les uson junto con operadores para vincularlos con los usuarios. 

### Usuarios
  * Propietarios: `u`
  * Grupo: `g`
  * Otros: `o`
  * Todas las clases: `a`

### Operadores
  * `+`: Añadir un permiso a la clase de usuario especificada o solo se sobreescriben
  * `-`: Eliminar un permiso de la clase de usuario especificada.
  * `=`: Establecer los permisos exactamente como se especifican, ignorando los permisos previos.


### Ejemplos
Con el comando `chmod` se usan estos operadores para cambiar permisos
  * `chmod u+x archivo.sh`: Otorga permisos de ejecución `(+u)` solo al propietario `(u)`
  * `chmod go-w archivo.txt`: Elimina permiso de escritura `(-w)` para el grupo `(g)` y otros `(o)`
  * `chmod a=rw archivo.txt`: Establece exactamente los permisos de lectura y escritura para todos `(a)`

## 2. Numérico u octal
Este permiso se trata de un número de tres cifras (XYZ) en el que cada posición representa una clase de usuario.

La notación octal sigue el mismo orden:
  * Primero (x): Es para el propietario
  * Segundo (y): Es para el grupo
  * Tercero (z): Es para otros

### Valores 
  * **Sin permisos (-)**: Valor es 0
  * **Leer (r)**: Valor es 4
  * **Escribir (w)**: Valor es 2
  * **Ejecutar (x)**: Valor es 1

De esto podrían surgir varias combinaciones
  * `0` -> Sin permisos
  * `1` -> Solo ejecución 
  * `2` -> Solo escritura
  * `3` -> Escritura y ejecución (2 + 1)
  * `4` -> Solo lectura
  * `5` -> Lectura y ejecución (4 + 1)
  * `6` -> Lectura y escritura
  * `7` -> Todos los permisos (4 + 2 + 1)

### Ejemplos
Se aplica también con `chmod`
  * `chmod 755 archivo`(rwxr-xr-x): Propietario tiene todos los permisos `7`, grupo `5` y otros `5` tienen permisos de lectura y ejecución.
  * `chmod 644 archivo`(rw-r--r--): Propierario lectura/escritura, grupo y otro solo tienen lectura 

### Permisos especiales `[especial] [usuario] [grupo] [otros]`
Tenenos permisos especiales como:
  * **SUID**(Set User ID):Representado por 4
  * **SGID**:(Set Group ID):Representado por 2
  * **Sticky bit**:Representado por 1
