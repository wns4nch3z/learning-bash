# TUBERÍAS | PIPES (|)
Un pipe conecta:
  * stdout (1) de un comando ==> stdin (0) del siguiente comando

```bash
comando1 | comando2
```
Significa

```less
[ comando1 ] --stdout--> | --stdin--> [ comando2 ]
```
> stderr NO pasa por el pipe, a menos que tú lo redirijas manualmente.



## Ejemplos

### Simple

````python
# echo produce stdout: "Hola Mundo"
# tr recibe esa cadena como stdin y la transforma
!echo "Hola Mundo" | tr a-z A-Z
````

### Stderr no pasa por el PIPE
```ruts
                (error)---------stderr----------> pantalla
[ ls /noexiste ] 
                (stdout vacío)---> | ---> (stdin) --> [ wc ] --> "0"

```

````python
!ls /noexiste | wc -l
````

### Stderr pasar por el PIPE
Hay que redirigirlo manualmente:
```bash
comando 2>&1 | otro
```
