# SUID | Set User ID

El `SUID` es un **bit especial** que se puede establecer en un archivo ejecutable. 

Cuando se aplica este bit, el archivo se ejecuta con los permisos **del propietario** del archivo, **no con los permisos del usuario que lo ejecuta**.

> Muy útil para tareas en las que se requiere ejecutar un programa con privilegios elevados, pero sin que el usuario tenga acceso total al sistema.


## ¿Cómo funciona?
  * Cuando un archivo tiene el bit **SUID** activado, cualquier usuario que ejecute ese archivo lo hace **con los privilegios del propietario** del archivo

## ¿Cómo verlo?
  * Si un archivo tiene el bit **SUID** activado, el primer carácter de los permisos en la salida de `ls -l` será una **`s`** en lugar de una `x` en el lugar del permiso del permiso de ejecución del **propietario**.
```bash
ls -l /usr/bin/passwd
# Salida
-rwsr-xr-x 1 root root 12345 dic 12 14:30 /usr/bin/passwd # Ejemplo de SUID
```

### `s`vs `S`
`SUID` **solo tiene sentido en archivos ejecutables**
Es decir debe tener permiso de ejecución `x`, no importa el orden pero debe quedar activada al final.

Con `x`:
```bash
-rwsr-xr-x 1 root root 12345 dic 12 14:30 /usr/bin/passwd 
```

Sin `x`:
```bash
# Representa son S mayúscula
-rwSr-xr-x 1 root root 12345 dic 12 14:30 /usr/bin/passwd 
```

> Sin `x`: No se ejecuta y SUID no funciona

## Asignaciones
### Simbólico
Con `sudo chmod u+s archivo`

### Númerico
Representado con el número 4 (4000)

`sudo chmod 4777 archivo` 

## Ubicar archivos SUID
```bash
find / -perm -4000 2>/dev/null
```
> `find` permite encontrar archivos


## NOTAS
  * Una explotación SUID NO es “explotar SUID” en abstracto
  * Es explotar un BINARIO mal diseñado que tiene SUID
  * La vulnerabilidad está en que el binario:
    * llama a otros programas (system, popen, exec*)
    * no valida rutas 
    * usa permisos incorrectos
    * hereda entorno
    * confía en input del usuario

El SUID no es la vulnerabilidad, es el amplificador

> Una explotación SUID es explotar un error de diseño en un programa que, debido a SUID, se ejecuta con más privilegios de los que debería.
