# WILDCARDS


### `*` | Cualquier cosa, cantidad
Muy usado para buscar archivos

```python
!ls ../files/borrador/
```

```python
!ls ../files/borrador/*.txt # Todos archivos txt
```

### `?` | Un solo carácter
Coincide con exactamente un carácter.

```python
!ls ../files/borrador/texto?.txt
```

### `[]`| Rangos o lista
Coinciden con uno de los caracteres listados dentro, o con un rango.

```python
!ls ../files/borrador/texto[1-2].txt
```

### `[! ] o [^ ]` | negación dentro de corchetes
Concide con cualquier carácter menos los indicados

```python
!ls [!2-3]*.ipynb
```

### `{}` | Brace expansion
No son exactamente wildcards, pero se usan mucho con ellos.

Permiten generar múltiples cadenas.
{a, b, c}.txt -> a.txt, b.txt, c.txt

> Genera cadenas de texto ANTES del globbing

```python
# Lista de valores
!ls ../files/borrador/{texto1,texto2}.txt
```

```python
# Lista números
!ls {2..4}*.ipynb
```
