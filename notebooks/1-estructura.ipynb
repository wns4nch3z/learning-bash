# Estructura de Directorios


Todo linux está organizado alrededor del directorio raíz `/`.

Debajo de él existen carpetas con funciones muy específicas.

Entender que contiene es fundamental para: 
  * Análisis forense
  * Privesec (escalada de privilegios)
  * Exploración de sistemas comprometidos
  * Configuración de persistencia
  * Deteccón de indicadores de compromiso

### 1. `/` | Directorio Raíz
Es la base del sistema. Todo cuega de aquí.
Navegar desde `/`te da la visión general del sistema.


```python
!tree -d -L 1 /
```

> Ver comando `tree` (Muestra como un arbol) para más detalles.

### 2. `/bin` | Programas esenciales
Contiene binarios básicos necesarios para que el sistema funciones:
  * bash
  * cat, cp, mv
  * ls
  * grep
  * entre otros.


### 3. `/sbin` | Binarios administrativos
Son comando de administrador: 
  * `iptables`
  * `reboot`
  * `fdisk`
  * `ifconfig`

> A veces los pentesters encuentran binarios en /sbin con permisos incorrectos o con SUID activado.

### 4. `etc` | Configuración del sistema
El directorio más valioso para auditorías/forense.
  * /etc/passwd -> usuarios (sin hashes de contraseña)
  * /etc/shadow -> hashes de contraseñas
  * /etc/sudoers -> permisos sudo
  * /etc/hosts -> resolución de nombres manual
  * /etc/crontab -> tareas programadas (útil para persistencia)
  * /etc/network -> configuración de red
  * /etc/ssh -> configuración de SSH, claves permitidas

### 5. `/home` | Directorios de usuarios
Aquí están archivos personales de cada usuario:
    * `/home/usuario/.ssh/autorized_keys`
    * `/home/usuario/.bash_history`
    * `/home/usuario/Documents`


### 6. `/root` | Home del usuario root
Aquí suele encontrarse:
  * Scripts de mantenimiento
  * Claves SSH del administrador
  * notas/credenciales

> Nota: Sólo existe en sistemas Linux (no en BSD).

### 7. `/var` | Archivos variables
Contiene cosas que cambian todo el tiempo
  * `/var/log` -> Registros del sistema 
    * `auth.log` -> intentos de login
    * `syslog` 
    * `apache2` -> logs de web
  * `var/www` -> Contenido web
  * `var/mail` -> correos locales

### 8. `/usr` | Programas y librerías de usuarios
Contiene la mayoría de programas instalados:
  * `/usr/bin` -> binarios de usuario (nmap, python, ssh, etc.)
  * `/usr/sbin` -> binarios admin
  * `/usr/lib` -> librerías
  * `/usr/local/` -> programas instalados manualmente

### 9. `/tmp` | Archivos temporales
Cualquiera puede escribir aquí.

Se borra automáticamente en reinicios.

Útil para:
  * Subir payloads sin sospecha
  * muchos exploits se ejecutan en `/tmp`

### 10. `/opt` | Programas instalados terceros
Ejemplo: 
  * Discord
  * Chrome
  * Aplicaciones no estándar

### 11. `/dev` | Dispositivos del sistema
  * `/dev/null`
  * `/dev/sda1` (discos)
  * `/dev/pts`(terminales)

### 12. `/proc` | Info del kernel y procesos
Directorio virtual (No ocupa espacio real):

Ejemplos:
  * `/proc/cpuinfo`
  * `/proc/meminfo`
  * `/proc/PID` -> Info de procesos


### 13. `/sys` | Info del sistema y hardware
Similar a `/proc`.

Se usa para interactuar con el kernel.
