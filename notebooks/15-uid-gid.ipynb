# UID / GID reales y efectivos
  * UID -> User ID (identificador del usuario)
  * GID -> Group ID (identificador del grupo)

Cada proceso en linux **siempre pertenece a un usuario y a un grupo**.

## 1. UID/GID real
Indica quién lanzó en proceso
```bash
usuario1$ nano archivo.txt
```
UID real = usuario1
GID real = grupo de usuario1
Sirve 

### 2. UID/GID efectivo
Indica con qué permisos se ejecuta el proceso

Esto es lo que el **kernel usa para permitir o negar acciones**

Ejemplo normal (sin SUID)
```bash
UID real = usuario1
UID efectivo = usuario1
```

### 3. ¿Cuándo son diferentes?
Cuando hay SUID o GUID

Ejemplos clásico: `/usr/bin/passwd`
```bash
ls -l /usr/bin/passwd
-rwsr-xr-x 1 root root ...
```
  * Dueño: `root`
  * SUID activo (`s`)

Cuando `usuario1` ejecuta `passwd`:
  * UID real: `usuario1`
  * UID efectivo: `root`

> El programa se ejecuta como root, aunque no seas root.

### 3. UID guardado (SUID - Saved User ID)
El UID “de respaldo”
  * Guarda el UID efectivo original
  * Permite:
    * Bajar privilegios temporalmente
    * Recuperarlos después de forma segura

### Frase para memorizar
  * Real = quién soy
  * Efectivo = qué puedo hacer ahora
  * Guardado = permiso que puedo recuperar
