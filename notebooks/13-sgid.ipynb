# SGID | Set Group ID
El **SGID** es otro bit especial que se puede establecer en archivos y directorios.

## Usos
Tiene dos usos principales dependiendo de si es un archivo o directorio.

### SGID en `archivos`
Cuando se establece en un archivo ejecutable, tiene un como¿portamiento similar al de **SUID**, pero con la diferencia de que **el programa se ejecuta con los privilegios del grupo del propietario del archivo**, no del propietario.

### SGID en `directorios`
Cuando se establece en un directorio, **todos los archivos creados dentro de ese directorio heredarán el grupo del directorio, no el grupo del usuario que crea el directorio**. 
> Útil para mantener coherencia dentro de un directorio compartido.


## ¿Cómo verlo?
Cuando un archivo o directorio tiene el bit **SGID** activado, se muestra con una **`s`** en lugar de **`x`** en el lugar del permiso de ejecución para el **grupo**.
### Ejemplo de **SGID en archivo**:

El comando `ls -l`: 
```
-rwxr-sr-x 1 root admin 12345 dic 12 14:30 /usr/bin/sgid_program
```

El **`s`** en el lugar de la **ejecución** del grupo indica que el bit **SGID** está activado.

> Similar la `S` mayúscula quiere decir que no esta activado la ejecución.

## Asignaciones
### Simbólico 
Con `sudo chmod g+s archivo/directorio`

### Numérico
Representado con el número 2 (2000)

Con `sudo chmod 2755 archivo`

## Ubicar archivos SGID
```bash
find / -perm -2000 2>/dev/null
```
