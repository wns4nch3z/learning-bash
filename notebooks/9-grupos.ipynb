# GRUPOS
Los grupos determinan **qué puede hacer un usuario**.

> Un usuario sin permisos pero en el grupo correcto = posible root

## ¿Qué es un grupo?
Un grupo es un conjunto de usuarios que comparten permisos sobre archivos, comandos o servicios.

## Ver a que grupos pertenezco

### Simple
`groups` o `id`
```bash
uid=1000(juan) gid=1000(juan) groups=1000(juan),27(sudo)
```
> `sudo` -> posible escalada

### En `/etc/group`
Cada línea de este archivo tiene la siguiente estructura:
```bash
nombre_de_grupo:contraseña:GID:lista_de_usuarios
``` 
  * Contraseña: Solía guardar contraseña ahora solo muestra una `x` o `*`, ahora se guarda en `/etc/gshadow`
  * Lista de usuarios: `juan, maria, pedro`

```bash
sudo:x:27:juan,maría,pedro
```
> Si encuentras usuarios que no deberían estar en grupos como sudo, admin, o wheel, podría ser una configuración incorrecta.
