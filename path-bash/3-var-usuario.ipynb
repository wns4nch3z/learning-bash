# VARIABLES
Son contenedores de información que pueden almacenar cadenas, números o resultados de comandos.

## 1. Asignación
### Sintaxis básica
```bash
# Definir
nombre_variable=valor
# Usarlo
$nombre_variable

```
Reglas importantes:
  * Los nombres no pueden comenzar con **un número ni contener espacios**.
  * Bash distingue entre mayúsculas y minúsculas: `miVariable`, `MiVariable` y `MIVARIABLE` son tres variables diferentes.
  * Si el nombre de la variable va seguido de otro carácter (letra, número o _), se debe usar `{}` para evitar ambigüedades: `${variable}_final`.

### Declaración
```bash
nombre="Ana"        # string
edad=30             # número (sigue siendo string)
activo=true         # bash no tiene booleanos reales
proyecto="mi app"   # con espacios requiere comillas
ruta=$(pwd)         # resultado de comandos
```

> Todo es string por defecto. Bash no distingue tipos como Python o JS. El número 30 y el string "30" son lo mismo para bash 

Formas de leer (expansión)
```bash
$nombre             # forma corta
${nombre}           # forma recomendada -delimita el nombre
"${nombre}"         # con comillas dobles - siempre así
"Hola ${nombre}, bienvenido" # interpolación en string
```

> Siempre citar con comillas dobles: `"${VAR}"`. Si la variable tiene espacios o está vacía y no la citas, bash la divide en palabras — un bug clásico que rompe rutas de archivos con espacios.

#### **Errores**
Declaración:
```bash
nombre = "Ana" # Bash intenta ejecutar el comando nombre
nombre="Ana"   # Sin espacios alrededor del =
```
Espacios en valor:
```bash
cp $archivo /tmp # Si hay espacios divide en dos argumentos
cp "${archivo}" /tmp/ # Siempre citar variables con rutas
```
Llaves en concatenación:
```bash
$archivo_bak   # Busca archivo_bak
${archivo}_bak # Llaves delimitan el nombre de la variable
```


## 

### Expansión con valores por defecto
```bash
# Si var está vacía o no existe
${VAR:-"default"}       # lee: usar VAR o "default"

# Si VAR esta vacía, asígnale "default"
${VAR:"default"}        # asigna si vacía

# Si var está vacía, muestra error y sale
${VAR:?"VAR es requerida"} # útil para validar inputs

# Ejemplo práctico:
entorno=${DEPLOY_ENV:-"development"}

```
> ${VAR:-default} es uno de los patterns más usados en scripts de DevOps — permite que una variable venga del entorno (CI, Docker) y si no está definida, usa un valor seguro.

## 2. Variables especiales
```bash
$0   # nombre del script
$1   # primer argumento
$2   # segundo argumento
$#   # cantidad de argumentos
$@   # todos los argumentos
$?   # código de salida del último comando
$$   # PID del script
```

Ejemplo: Al ejecutar el script `./scritp.sh` se puede pasar argumentos `./script argumento1 argumento2 ... argumentoX`

```bash
#!/bin/bash

echo "Nombre: $0"
echo "Soy el primer: $1"
echo "Soy el segundo: $2"

echo "Cantidad de argumentos: $#"
echo "PID del script: $$"

```
Salida:
```ouput
Nombre: ./argumentos.sh
Soy el primer: uno
Soy el segundo: dos
Cantidad de argumentos: 2
PID del script: 4214
```

## 3. Constantes
La palabra clave readonly se utiliza para definir constantes, es decir, variables cuyo valor no puede modificarse ni eliminarse una vez declaradas.

Uso básico
```bash
readonly PI=3.14159
readonly MAX_USERS=100
```

Una vez declarada como `readonly`, cualquier intento de modificar o eliminar la variable produce un error:
```bash
PI=3.14     # Error: bash: PI: readonly variable
unset PI    # Error: bash: unset: PI: cannot unset: readonly variable
```

## 4. Ejercicio

### Objetivo

Crear un script que:
  * Use variables
  * Use expansión ${}
  * Use valores por defecto

***Enunciado:***
Crear un script que reciba 3 variables:
  1. Reciba 3 variables
    * `nombre`
    * `país`
    * `edad`

  2. Si alguna no está definida, usar por defecto
    * nombre -> `"Invitado"`
    * país   -> `"Desconocido"`
    * edad   -> 18 

Mostrar mensaje como: `Hola Juan, tienes 25 años y vives en Peru`




````python
%%bash
#!/bin/bash

nombre=$1
pais=$2
edad=$3

nombre_final="${nombre:-Invitado}"
pais_final="${pais:-Desconocido}"
edad_final="${edad:-18}"

echo "Hola $nombre_final, tienes $edad_final años y vives en $pais_final"
````
