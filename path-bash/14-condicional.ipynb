# Condicional `if`
Es una condición que permite ejecutar comandos solo si algo es verdadero en este caso su salida debe ser éxito (0).

## Sintaxis
### Básica
```bash
if [ condicion ]; then
    comando
fi
```
Importante:
  * `[ ]` es obligatorio en la forma clásica
  * `then` va después de la condición
  * `fi` es `if` al revés (cierra el bloque)

````python
%%bash
edad=18

if [ $edad -ge 18 ]; then
    echo "Eres mayor de edad"
fi
````

---

### `if` con `else`
Si la condición no es exitosa se usa `else` para ejecutar otras líneas de comandos.

> Es opcional y debe ser el último bloque de la estructura. 

```bash
if [ condicion ]; then
    comandos
else
    comandos
fi
```


````python
%%bash
edad=15

if [ $edad -ge 18 ]; then
    echo "Mayor de edad"
else
    echo "Menor de edad"
fi
````

---

### `if`, `elif` y `else`
Se utiliza para probar una nueva condición si la condición del if (o un elif anterior) fue falsa.

Se pueden encadenar múltiples bloques elif. El primer `elif` cuya condición se evalúe como verdadera ejecutará su bloque de código y el flujo saldrá de toda la estructura `if`, ignorando las condiciones restantes.

```bash
if [ condición ]; then
    comandos
elif [ condicón ]; then
    comandos
elif [ condición ]; then
.
.
.
else
    comandos
fi
```


````python
%%bash
nota=15
if [ $nota -ge 18 ]; then
    echo "Excelente"
elif [ $nota -ge 13 ]; then
    echo "Aprobado"
else 
    echo "Desaprobado"
fi
````

## Formas
### Corta 
```bash
[ $edad -ge 18 ] && echo "Mayor de edad"
```
Otra:
```bash
[ $edad -lt 18 ] || echo "Es mayor o igual a 18"
```

### Ejemplos

````python
%%bash
# NUMEROS
edad=15
if [ $edad -lt 14 ]; then
    echo "Entrando xd"
fi
````

````python
%%bash
# STRINGS
saludo="Hola"

if [ $saludo = "Hola" ]; then
    echo "Es igual"
else
    echo "No es igual"
fi
````

````python
%%bash
# ARCHIVO

if [ -e 0-shell.ipynb ]; then
    echo "Existe"
fi
````

## Recomendación
Es más seguro usar `[[ ]]`
  * Es más segura con variables vacías
  * Permite regex
  * No necesita comillas
