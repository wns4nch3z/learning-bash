# REGEX
Una regex es un **patrón para buscar texto**.

> No comparas texto exacto, sino una **regla**.

Ejemplo:
  * `"Juan"` -> texto exacto
  * `^J` -> cualquier texto empiece con J

## Operador `=~`
Es un operador de comparación con expresiones regulares (regex)
```bash
[[ texto =~ regex ]]
```

> En regex no usar entre comillas sino lo tratará de forma literal

--- 


## Patrones

### `.` | El punto

Se interpreta como cualquier carácter

```bash
[[ "hola" =~ h.la ]]
```
Coincide con `hola`, `h1la`, `h-la` 

> `.` = cualquier carácter (solo uno)


````python
%%bash
[[ "h1la" =~ h*la ]] # éxito | true | 0
echo $?
````

### `*` | Asterisco
No dice **cero o más** repeticiones del elemento anterior.

```bash
[[ "hoola" =~ ho*la ]]
```
> Significa `o*`: cero o más `o`

Coincide con: `hola`, `hoola`, `hoooola`


````python
%%bash 

[[ "hla" =~ ho*la ]]
echo $?
````


### `+` | El signo más
Dice **uno o más** repeticiones del elemento anterior.

```bash
[[ "hola" =~ ho+la ]]
```
> Significa `o+`: al menos una `"o"`

Coincide con `hoola`, `hoola`, no con `hla`

````python
%%bash
[[ "hla" =~ ho+la ]] # al menos uno debería
echo $?
````

### `?` | Interrogación
El `?` quiere decir **0 o 1 vez** (opcional)

```bash
[[ "color" =~ colou?r ]]
```
Significa `u?`: puede haber `u` o no.

Coincide con `color`, `colour`

````python
%%bash

[[ "abc" =~ ab?c ]] # abc / ac
echo $?
````

| Símbolo | Significado |
| ------- | ----------- |
| `*`     | 0 o más     |
| `+`     | 1 o más     |
| `?`     | 0 o 1       |

----

### `[]` | Corchetes
Representa conjunto

```bash
[[ "hola" =~ h[oa]la ]]
```
Significa `[oa]`: puede ser "o" o "a"

Coincide con `hola`, `hala`

````python
%%bash
[[ "hala" =~ h[oa]la ]]
echo $?
````

### `[x-y]` | Rangos

```bash
[[ "h3la" =~ h[0-9]la ]]
```
Siginifca `[0-9]`: cualquier número dentro del rango

Coincide con `h1la`, `h2la`

Formas:
  * [0-9]
  * [a-z]
  * [A-Z]

````python
%%bash
[[ "h9la" =~ h[0-9]la ]]
echo $?
````

### `^` y `$`
Representa inicio (`^`) y fin (`$`) 
```bash
[[ "hola" =~ ^h ]]
```
Empieza con `"h"`

```bash
[[ "hola" =~ a$ ]]
```
Termina con `"a"`

````python
%%bash
[[ "hola" =~ ^h ]]
echo $?
````

````python
%%bash
[[ "hola" =~ a$ ]]
echo $?
````

### `[^ ]`
Esto se interpreta como ***negación*** -> "todo menos esto"

```bash
[[ "hola" =~ h[^o]la ]]
```
Significa `[^o]`: cualquier cosa menos `o`

Coincide con: `h1la`, `hala`, pero no con `hola`

Forma mental:
  * `[abc]` → uno de estos
  * `[^abc]` → cualquiera menos estos

### Escapar
Para escapar caracteres (`\`)
```bash
[[ "file.txt" =~ \.txt$ ]]
```

  * `\.` = punto literal
  * sin `\` sería cualquier carácter

--- 

## Ejemplo

````python
%%bash
# Solo números
[[ "123" =~ ^[0-9]+$ ]]
echo $?
````

````python
%%bash
# Solo letras y números 
[[ "abc123" =~  ^[a-zA-Z0-9]+$ ]]
echo $?
````

````python
%%bash
# Digitos exactos 3
[[ "456" =~ ^[0-9]{3}$ ]]
echo $?
````

````python
%%bash
# Empiece con "a" y termina "z"
[[ "abcz" =~ ^a.*z$ ]]
echo $?
# ".*" -> cualquier cosa en medio o incluyendo nada
````

````python
%%bash
# Email Simple
[[ "usuario123@gmail.com" =~ ^[a-z0-9]+@gmail\.com$ ]]
echo $?
````

````python
%%bash
# IP SIMPLE
[[ "192.168.1.1" =~ ^[0-9]+\.[0-9]+\.[0-9]+\.[0-9]+$ ]]
echo $?
````

````python
%%bash
# Usuario entre 4-10 caracteres
# Permite solo letras y números
[[ "willian" =~ ^[a-zA-Z0-9]{4,10}$ ]]
echo $?
````

````python
%%bash
# Contraseña simple
contra="9385"
[[ $contra =~ [0-9] && $contra =~ ^.{6,}$ ]]
echo $?
````

## Match parcial vs Validación completa

### Caso base
```bash
[[ "abc123" =~ [^0-9]+ ]]
```
Resultado: true

Significa: **"uno o más cosas que no son números"**

Y en `abc123` existe `abc` entonces si encuentra coincidencia.

### Match parcial
Bash piensa así:

> "¿hay alguan parte del texto que cumpla esto?"

si -> entonces TRUE

```txt
abc123
^^^
match aquí
```

### Validación completa
Ahora:
```bash
[[ "abc123" =~ ^[^0-9]+$ ]]
```
Resultado: FALSE

Ahora dice `^`(inicio), `$` (fin) o sea **TODO el texto debe cumplir la condición**.

```txt
abc123
   ^^^ ← números
```
Rompe la regla

> Si quieres validar, SIEMPRE usa: `^ ... $`

Diferencia:

| Tipo     | Regex      | Qué hace                 |
| -------- | ---------- | ------------------------ |
| Parcial  | `[a-z]+`   | busca en cualquier parte |
| Completo | `^[a-z]+$` | valida TODO el texto     |
