Un file descriptor (fd) es simplemente un número entero que el kernel asigna a un canal de comunicación abierto. Todo en Unix es un archivo — un socket, un pipe, un terminal, un archivo real — y se accede a través de un fd. Los tres primeros (0, 1, 2) los crea el kernel automáticamente para cada proceso.

## FDs en scripts

### Script básico

````python
%%bash
#!/bin/bash

echo "Es es una salida normal"
echo "Esto es un error" >&2 # Redirigir el stdout a stderr usando FDs 2 
````

---

### Redirigir todo

```bash
#!/bin/bash

exec > salida.log 2> errores.log

echo "Todo lo que salga va a los logs"
ls /noexiste
```

> `exec` cambia los FDs **del proceso actual**, no de un solo comando.

### Creación de FDs personalizados
FDs exclusivos para logs:


```bash
#!/bin/bash

exec 3> debug.log # Agregando FDs

echo "Inicio del script" >&3
echo "Salida normal"
ls /noexiste >&3

exec 3>&- # Cerrando
```

Ventaja:

* stdout → para el usuario
* FD 3 → para debug/log
