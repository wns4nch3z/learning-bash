# Variables de entorno
Las variables de entorno en Linux son parámetros que almacenan información utilizada por el sistema y las aplicaciones, como rutas de acceso, configuraciones regionales, el nombre de usuario o la zona horaria. 

Estas variables permiten personalizar el comportamiento del sistema y facilitan la comunicación entre procesos.

## 1. Tipos de variables de entorno

### Variables de Shell
Solo existen durante la sesión actual y no se heredan por procesos hijos. Ejemplo:
```bash
mi_var="valor"
```

### Variables de entorno:
Se heredan por procesos hijos y están disponibles en la sesión actual. Se declaran con `export`: `export VAR='valor'`


### Variables permanentes:
Sobreviven al cierre de sesión. Se configuran en archivos como `~/.bashrc`, `~/.profile`, `~/etc.environment`


## 2.Variables comunes
  * `PATH`: Define los directorios donde se buscan comandos ejecutables.
  * `HOME`: Ruta al directorio principal del usuario
  * `PWD`: Directorio de trabajo actual
  * `USER`: Nombre del usuario actual
  * `SHELL`: Shell predeterminado del usuario
  * `PS1`: Formato del prompt de la terminal
  * `LANG`: Idioma para mensajes del sistema
  * `TZ`: Zona horaris
  * `DISPLAY`: Pantalla para mostrar aplicaciones gráficas
  * `OLDPWD`: Directorio anterior (antes del último cd)

### 3. Como gestionarlas

- **Listar variables**: Usa `printenv`, `env` o `set`.
- **Crear una variable temporal**: `export MI_VARIABLE="valor"`.
- **Crear una variable permanente**: Edita `~/.bashrc` o `/etc/environment` y agrega `export MI_VARIABLE="valor"`, luego ejecuta `source ~/.bashrc` o `source /etc/environment`.
- **Ver el valor de una variable**: `echo $MI_VARIABLE`.
- **Eliminar una variable**: `unset MI_VARIABLE`.







### 4. Archivos clave
- `~/.bashrc`: Para variables locales del usuario.
- `~/.profile` o `~/.bash_profile`: Para variables de sesión remota.
- `/etc/environment`: Para variables globales para todos los usuarios.
- `/etc/profile`: Para configuraciones del sistema al iniciar sesión.

### 5. HERENCIA
  * **SO/KERNEL**: Define variables base del sistema. Las hereda el shell al arrancar.
  * **SHELL**: Tiene sus propios vars locales + las heredadas del SO. Solo pasa a hijos las que tiene export.
  * **SCRIPT**: Recibe las vars exportadas del shell padre. Sus vars locales (sin export) no llegan a procesos.
  * **PROCESO HIJO**: Solo ve lo que el script exportó. No puede modificar las vars del padre - los cambios son en su copia.

> **Los procesos hijos reciben una copia del entorno, nunca una referencia**. Si un script hijo modifica $DB_HOST, el padre no lo ve. La comunicación de hijo a padre se hace por stdout, archivos o códigos de salida — no por variables.  

```bash


TOKEN="abc123"            # variable local — hijos NO la ven
export DB_URL="postgres://…"  # exportada — hijos SÍ la ven

# Exportar y asignar en una línea
export NODE_ENV="production"

# Pasar una variable solo para un comando (no persiste)
DEBUG="true" ./mi-script.sh
```
