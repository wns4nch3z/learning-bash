# READ
`read` lee desde stdin (FD 0) y guarda el valor en una variable.

```bash
# Ejemplo básico
read nombre
echo "Hola $nombre"
```

>  El script se queda esperando a que el usuario escriba algo, guarda y después muestra según código.

## Opciones

### `-p`
Leer con mensajes
```bash
read -p "Ingresa tu nombre: " nombre
echo "Hola $nombre"
```

### `-s`
Leer sin mostrar (contraseñas)
```bash
read -s -p "Password: " pass
echo
echo "Guardado"
```
> `-s` → silencioso (no muestra lo que escribes)

### `-t`
Leer con timeout
```bash
read -t 5 -p "Tienes 5 segundos: " dato
echo "Ingresaste: $dato" 
```
> si no se escribe nada continúa

### Leer múltiples variables
```bash
read nombre edad
echo "Nombre: $nombre"
echo "Edad: $edad"
```
Entrada
```
Juan 20
```

### Leer desde archivos (sin loops)
Usando redirección:
```bash
read linea < archivo.txt
echo $linea
```
> Esto lee solo la primera línea


### Leer desde otro comando (FDs)
```bash
read linea < <(echo -e "Línea 1\nLínea 2")
echo $linea
```

  * `<(...)` crea un flujo (como archivo temporal)
  * `read` lo consume desde stdin


### Usar FDs explícitamente con read
```bash
exec 3> archivo.txt

read linea <&3
echo "$linea"

exec 3>&- # Cerrando FD 3
```

### Leer un carácter
```bash
read -n 1 -p "Presiona una tecla: " tecla
echo 
echo "Presionaste: $tecla"
```
