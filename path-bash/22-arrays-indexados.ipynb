# ARRAYS ASOCIATIVOS
Es una estructura de datos que permite almacenar múltiples valores en una sola variable, donde cada elemento se accede mediante un índice numérico (empezando desde 0).

## Sintaxis

### Declarar un array
```bash
mi_array=(valor1 valor2 valor3)
```

### a. Acceder a elementos
Accede a través de índices empieza desde cero
```bash
# Acceder a elemento 1
${mi_array[0]} 
# Acceder a elemento 2
${mi_array[1]}
# Acceder a elemento 3
${mi_array[2]}
```

### b. Mostrar todos el array
```bash
# Mostrando
${mi_array[@]}
```

### c. Mostar cantidad de elementos
```bash
${#mi_array[@]}
```

### d. Agregar elementos
```bash
mi_array+=("valor4")
```

### e. Modificar elemento
```bash
mi_array[1]="valorvip"
```

### f. Eliminar elemento
```bash
unset mi_array[1]
```

### g. Ver índices
```bash
${!mi_array[@]}
```

### h. Copiar arrays
```bash
original=("a" "b" "c")

copia=("${original[@]}")

```

### i. Unir arrays
```bash
a=("uno" "dos")
b=("tres" "cuatro")

c=("${a[@]}" "${b[@]}")
```

````python
%%bash
# Ejemplo
nombres=("Willian" "Juan" "Luis")

printf "Accediendo\n"
printf "Elemento 1: %s\n" "${nombres[0]}"
printf "Elemento 2: %s\n" "${nombres[1]}"
printf "Elemento 3: %s\n" "${nombres[2]}"

printf "\nMostrando todo\n"
printf "Elemento -> %s\n" "${nombres[@]}" # Muestra como un for

printf "\nModificando\n"
nombres[1]="Marco"
printf "Elementos -> %s\n" "${nombres[*]}" # Muestra todo en una sola cadena

printf "\nEliminar elemento\n"
unset nombres[1]
printf "Elementos -> %s\n" "${nombres[*]}" 

printf "\nVer índices\n"
printf "Indice: %s\n" "${!nombres[@]}"

printf "\nCantidad \n"
printf "Número de elementos: %s" "${#nombres[@]}"

````

````python
%%bash
# Agregando dinámicamente
paises=()

paises[0]="Perú"
paises[1]="Brasil"
paises[2]="Chile"

printf "País -> %s\n" "${paises[@]}"
````

````python
%%bash
# Usando índice y valor con for
valores=("a" "b" "c")

# For simple
printf "\nMostrando índices\n"
for indice in "${!valores[@]}"
do
  printf "Índice: %d\n" "$indice"
done

# Combinando
printf "\nCombinando\n"
for indice in "${!valores[@]}"
do 
  printf "Índice: %d | Valor %s\n" "$indice" "${valores["$indice"]}" 
done
````

### Leer archivos
En Bash, `readarray` (también conocido como `mapfile`) se usa para leer líneas de un archivo o entrada estándar y guardarlas en un array. Es muy útil cuando quieres procesar archivos línea por línea sin usar bucles manuales.

````python
%%bash
# Leer
readarray -t rutas < ../files/borrador/lista.txt

for linea in "${!rutas[@]}"
do
  printf "Línea %d -> %s\n" "$linea" "${rutas[$linea]}"
done

````

````python
%%bash
readarray -t rutas < ../files/borrador/lista.txt

for indice in "${!rutas[@]}"
do
  linea="${rutas[$indice]}"  

  # Ignorar vacías y comentarios
  if [[ "$linea" =~ ^[[:space:]]*$ || "$linea" =~ ^[[:space:]]*# ]];then
    continue
  fi

  # Controlar archivos sin extensión
  archivo="${linea##*/}"
  
  if [[ "$archivo" == *.* ]];then
    extension="${archivo##*.}"
    nombre="${archivo%.*}"
  else 
    extension=""
    nombre="$archivo"
  fi
  
  # Mostrar
  printf "[%d] %s (%s) \n" "$indice" "$nombre" "$extension"

done 
````
