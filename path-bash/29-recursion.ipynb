# Recursión

## ¿Qué es la recursión?
Es cuando una función se llama a sí mismo hasta umplir una condición de parada.

Siempre necesita:
  * **Case base** (para detener la recursión)
  * **Caso recursivo** (donde la función se vuelve a llamar)

### Ejemplo Factorial
El factorial de un número es:
```txt
5!=5x4x3x2x1
```

Definición recursiva:
```txt
n! = n x (n-1)!
```
Caso base:
```txt
0! = 1
```
#### Como piensa la recursión
Si llamas `factorial 4`, pasa esto:
```bash
factorial(4)
= 4 * factorial(3)
= 4 * (3 * factorial(2))
= 4 * 3 * (2 * factorial(1))
= 4 * 3 * 2 * (1 * factorial(0))
= 4 * 3 * 2 * 1 * 1
```

#### Implementación en bash

````python
%%bash
# Factorial
factorial() {
    local n=$1

    if [ "$n" -le 1 ]; then
        echo 1
    else
        local prev=$(factorial $((n - 1)))
        echo $((n * prev))
    fi
}

factorial 5
````

#### Cosas importantes
  * Bash no está diseñado para recursión pesada
  * Puede volverse lento o romperse con números grandes
  * Cada llamada crea un subshell → costo alto


### Ejemplo de recursión de árbol de directorios
Aquí es donde la recursión se vuelve súper útil en la vida real.

Un sistema de archivos es un árbol:
```txt
/home
 ├── user
 │    ├── docs
 │    └── fotos
 └── etc
```

#### Idea clave 
Un directorio puede contener:
  * archivos
  * otros directorios

Entonces:

> “Para listar todo, haz lo mismo dentro de cada subdirectorio”

Eso es recursión pura.

```bash

```

````python
%%bash
listar() {
    local dir="$1"

    for archivo in "$1"/*
    do
        echo "Archivo: $archivo"

        if [[ -d "$archivo" ]]; then
            listar "$archivo"
        fi
    done
}

listar "../files/borrador"
````

#### Flujo
  1. Empieza en un directorio
  2. Listar todo
  3. Si encuentra otro directorio -> se llama a sí misma
  4. Repite hasta que ya no hay más carpetas


#### Problemas comunes

##### a. Directorios vacíos
El glob `*` puede fallar -> usar:
```bash
shopt -s nullglob
```

#### b. Enlaces simbólicos
Podría entrar en ciclos

Solución:
```bash
[[ -L "$archivo" ]] && continue
```

#### c. Profundidad infinita
Puedes limitar:
```bash
listar() {
    local dir="$1"
    local depth="$2"

    [ "$depth" -le 0 ] && return

    for archivo in "$dir"/*; do
        echo "$archivo"

        if [ -d "$archivo" ]; then
            listar "$archivo" $((depth - 1))
        fi
    done
}
```

#### Consejo
Aunque aprender recursión en Bash es útil:

En la práctica, se usa más:
  * `find`
  * `tree`
  * `ls -R`
