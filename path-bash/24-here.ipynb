# HERE
En Linux (bash y shells compatibles), los heredocs y here-strings sirven para enviar texto directamente a la entrada estándar (stdin) de un comando sin crear archivos temporales.

## `here-doc`

### `<< EOF` - here-doc básico
Envía varias líneas como entrada a un comando permitiendo expansión de variables y comandos.

```bash
cat << EOF
Hola $USER
Hoy es $(date)
EOF
```
Resultado:
  * $USER se reemplaza por tu usuario
  * $(date) se ejecuta

### `<< 'EOF'` - here-doc literal
Todo el contenido se trata como **texto puro**, sin expansión
```bash
cat << 'EOF'
Hola $USER
Hoy es $(date)
EOF
```
Resultado:
  * Se imprime literalmente $USER y $(date)

Útil para:
  * Scripts
  * Código con $, backticks, etc.


### `<<- EOF` - here-doc con indentación
Permite usar `tabs`(no espacios) antes del contenido y del delimitador final.
```bash
cat <<- EOF
	Linea con tab
	Otra linea
EOF
```
> Bash eliminará los tabs iniciales, útil para mantener el código ordenado.

## `<<<` - here-strings
Envía **una sola línea** como `stdin`.

```bash
echo "hola mundo" | grep "hola"
```
> Más corto y directo


Cuándo usar cada uno:
  * `<< EOF` → cuando quieres expansión
  * `<< 'EOF'` → cuando quieres texto literal
  * `<<- EOF` → cuando quieres indentar bonito
  * `<<<` → cuando solo necesitas una línea rápida

> Delimitador puede ser cualquier palabra no necesariamente tiene que ser `EOF`

## Usos
### Guardar en archivo
```bash
cat << EOF > archivo.txt
Linea 1
Linea 2
EOF
```

### Append
```bash
cat << EOF >> archivo.txt
Otra linea
EOF
```

### Scripts 
```bash
#!/bin/bash

usuario="admin"

cat << EOF
Creando usuario: $usuario
EOF 
```
