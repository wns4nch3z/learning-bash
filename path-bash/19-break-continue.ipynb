# CONTROL DE FLUJO

## BREAK
Rompe el bucle completamente.

Sale del `for`, `while` o `until` inmediatamente.

## CONTINUE
Salta a la siguiente iteración.

No termina el bucle, solo ignora lo que queda y vuelve a evaluar la condición.

## Ejemplos

````python
%%bash
# Usando break
for i in {1..10}
do 
  if (( i == 5 )); then
    echo "Se rompre en $i"
    break # Al llegar sale del bucle
  fi
  echo "Número $i"
done
````

````python
%%bash
# Usando continue
for i in {1..5}
do 
  if (( i == 3 )); then
    continue # No ejecuta la siguiente línea, salta a la siguiente iteración
  fi
  echo "Número: $i"
done
````

````python
%%bash
# Usados en until
contador=0

until (( contador > 5 ))
do 
  (( contador++ ))

  if (( contador == 3 )); then
    echo "Saltando"
    continue
  fi
  if (( contador == 5 )); then
    echo "Terminando"
    break
  fi 
  echo "Valor $contador"
done
````

````python
%%bash
# Bucles anidados
for i in {a..c}
do 
  for j in {1..3}
  do 
    if (( j == 2 ));then
      break
    fi
    echo "i=$i | j=$j"
  done
done
````

También se pueden usar con niveles (en bucle anidados)
```bash
