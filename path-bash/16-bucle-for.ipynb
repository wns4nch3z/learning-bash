# BUCLE FOR
Un `for` repite un bloque de código varias veces, recorriendo una lista de valores.

---

## Sintaxis 
```bash
for variable in lista_de_valores
do
    comandos
    comandos
    # ...
done
```
En una línea:
```bash
for i in lista; do comando; done
```
  * `variable`: Nombre de la variable que toma cada valor de la lista.
  * `lista_de_variables`: Puede ser una lista explícita, un rango numérico, una expansión de comodines (como `*.txt`) o la salida de un comando.
  * `do` y `done`: delimitan el bloque de comandos a ejecutar.


---


````python
%%bash
for i in 1 2 3 4 5;do
    echo "$i"
done
````

## Formas

### Rangos `{}`
Bash permite generar secuencias automáticas.

````python
%%bash
for i in {1..5}
do
    echo "$i"
done
````

### Glob

````python
%%bash
for i in [0-9]-*
do
    echo "$i"
done
````

### Líneas de un comando

````python
%%bash
# seq genera secuencia de números, cada valor en nueva línea
for i in $(seq 1 10 )
do
    echo "Soy el $i"

done
````

### Sintaxis tipo C | Contadores
Útil para iterar un número específico de veces.

Similar a `C`, `Java`, etc.

```bash
for (( inicialización; condición; incremento ))
do
    comandos
done
```

````python
%%bash
for (( i=1; i<=5; i++ ))
do 
    echo "Línea: $i"
done
````
