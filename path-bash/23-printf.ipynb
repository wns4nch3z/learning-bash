# PRINTF

## ¿Por qué usar `printf` en vez de `echo`?
`echo` parece simple, pero tiene problemas:
```bash
echo "-n Hola"
echo "\n Hola"
```
Dependiendo del sistema puede:
  * Interpretar `-n` o no
  * Ignorar `-n` o no
  * Romper scripts portables

`printf` es:
```txt
✔ Predecible  
✔ Portable  
✔ Profesional  
✔ Permite formatear salida  
```

## Uso 

### Básico
```bash
printf "Hola mundo"
```

> Nota: No agrega salto de línea automáticamente por se usa `\n`.

````python
%%bash
# Sin salto
printf "Hola "
printf "Mundo, "

# Con Salto
printf "Saltando\n"
printf "En otra línea"
````

### Variables
Definiendo tipo de dato que se está imprimiendo:
| Formato | Significado        | Ejemplo |
| ------- | ------------------ | ------- |
| `%s`    | string (texto)     | "Hola"  |
| `%d`    | entero             | 25      |
| `%f`    | flotante (decimal) | 3.14    |
| `%c`    | carácter           | 'A'     |

Una variable

````python
%%bash
nombre="Willian"
printf "Nombre: %s" "$nombre"
````

Múltiples variables

````python
%%bash
nombre="Juan"
edad=20
ciudad="Lima"

printf "Mi nombre es %s, tengo %d años y soy de %s.\n" "$nombre" $edad "$ciudad"
````

````python
%%bash
PI=3.14159

printf "PI con 2 decimales: %.2f\n" $PI
````

### Formatos
Idea clave:
```bash
printf "FORMATO" valores...
```
El formato es un plantilla donde defines:
  * que tipo de dato esperas
  * cómo quieres que se vea (ancho, decimales, etc.)


#### ***Estructura***
Cada formato empieza con `%` y sigue esta estructura:
```bash
%[flags][ancho][.precisión][tipo]
```
Ejemplo:
```bash
%-10.2f
```
##### Tipo
Los tipos como `s` (string), `%d` (entero), `%f` (flotante), `%c` (carácter)

##### Ancho (width)
Define el mínimo de espacio que ocupará el valor.
```bash
%10s
```
> mínimo 10 caracteres

````python
%%bash
printf "X%10sX\n" "Ana"
````

##### Flags (banderas)
Controlan alineación y relleno.

**`-` (alinear a la izquierda)**

````python
%%bash
printf "X%-10sX\n" "Ana"
````

**`0` (rellenar con ceros)**

````python
%%bash
printf "%05d\n" 123
````

**`.` (Precisión)**

Depende del tipo:

  * Números (`%f`)

````python
%%bash
# Números limitar decimales
printf "%.2f\n" "3.14159"
````

  * Para strings (`%s`)

````python
%%bash
# Strings máximo de caracteres
printf "%.3s" "Carlos"
````

##### Combinaciones
Ejemplo completo

````python
%%bash
printf "%-10.3s %08.2f\n" "Carlos" 3.5
````

Cada `%` consume un valor:
```bash
printf "%s %d %f\n" "Ana" 20 3.14
```
  * Si faltan valores -> comportamiento raro
  * Si sobran -> se reutilizan el formato

````python
%%bash
# Ejemplo curioso
printf "%s %d\n" "Ana" 20 "Luis" 30
# Reutiliza el formato automáticamente
````
