# GLOB PATTERNS
Los GLOB patterns en Bash son patrones de coincidencia de nombres de ruta (wildcards) que permiten al shell expandir automáticamente una expresión para coincidir con nombres de archivos, directorios o enlaces simbólicos existentes en el sistema.

## 1. Básicos

### `*`
Coincide con **cero o más caracteres**

> `*` = **cualquier cosa**

```python
%%bash
[[ "abc123" == * ]] 
echo $? 
```

### `?`
Coincide con exactamente un carácter

```python
%%bash
[[ "a-b" == a?b ]]
echo $?
```

### `[...]`
Coincide con un solo carácter dentro de los corchetes por ejemplo, `[abc]` o `[a-z]`

```python
%%bash
[[ "hpla" == h[ab]la ]]
echo $?
```

```python
%%bash
[[ "hola" == h[a-m]la ]]
echo $?
```

### `[!...]` o `[^...]` 
Coincide con un carácter que no esté en la lista

```python
%%bash
[[ "file.txt" == file.[!t]* ]] # Primera letra de extensión no sea t
echo $?
```

### Otros usos `*`
#### `a*`
Empieza con a

```python
%%bash
[[ "abc" == a* ]]
echo $?
# Empieza con "a" + cualquier cosa después
```

#### `*a`
Termina en `a`

```python
%%bash
[[ "casa" == *a ]]
echo $?
# Cualquier cosa + termina en "a"
```

#### `*a*`
Contiene "a"

```python
%%bash
[[ "perro" == *e* ]]
echo $?
# * → cualquier cosa antes
# contiene al menos una "e"
# * → cualquier cosa después
```

## Patrones extendidos
Requieren activarse con `shopt -s extglob`. Incluyen operadores similares a las expresiones regulares (regex).

  * `?(patrón)`: cero o una ocurrencia del patrón
  * `*(patrón)`: cero o más ocurrencias 
  * `+(patrón)`: una o más ocurrencias
  * `@(patrón)`: exactamente una ocurrencia (como una lista de opciones) 
  * `!(patrón)`: cualquier cosa **excepto** el patrón

## SHOPT
El comando shopt en Bash permite habilitar o deshabilitar opciones específicas que modifican el comportamiento de la shell. Se utiliza con las opciones `-s` (set) para activar y `-u` (unset) para desactivar. 

  1. `extglob`: Habilita los patrones de coincidencia extendidos como *(patrón), +(patrón), etc.
  2. `dotglob`: Incluye archivos ocultos (que comienzan con .) en la expansión de nombres de archivo. 
  3. `globstar`: Permite que el patrón `**` coincida con directorios de forma recursiva.
  4. `nullglob`: Si un patrón no coincide con ningún archivo, se expande a una cadena vacía.
  5. `failglob`: Si un patrón no coincide con ningún archivo, se muestra un error.
  6. `autocd`: Permite cambiar al directorio simplemente escribiendo su nombre (en shells interactivos).
  7. `cdspell`: Corrige errores tipográficos menorees en los nombres de directorios al usar `cd`.
  8. `cmdhist`: guarda comandos de varias líneas como una sola entrada en el historial.
