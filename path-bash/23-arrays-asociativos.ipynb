# Array Asociativo
Es un array donde:
  * Cada valor está asociado a una **clave**
  * Las claves suelen ser texto
  * No hay orden garantizado

Ejemplo mnetal:
```bash
usuario["nombre"]="Juan"
usuario["edad"]=25
```

## Sintaxis
Se debe declarar explícitamente:

```bash
declare -A mi_array
```

> Si no haces esto, bash asumirá como un array normal (indexado).

### a. Asignación de valores
```bash
mi_array["nombre"]="Luis"
mi_array["edad"]=20
mi_array["pais"]="Perú"
```

También puedes inicializar en una sola línea:
```bash
declare -A usuario=(
  ["nombre"]="Ana"
  ["edad"]=28
  ["ciudad"]="Trujillo"
)
```

### b. Acceder a valores
```bash
${mi_array["nombre"]}
${mi_array["edad"]}
```

### c. Mostrar claves
```bash
${!mi_array[@]}
```

### d. Mostrar valores
```bash
${mi_array[@]}
```

### e. Añadir 
```bash
mi_array["correo"]="luisito@gmail.com"
```

### f. Modificar
```bash
mi_array["correo"]="luispro@gmail.com"
```

### g. Eliminar 
Específico
```bash
unset mi_array["correo"]
```
Eliminar todo
```bash
unset mi_array
```

### NOTAS

#### Claves no pueden ser:

  * Vacías (`""`) -> **error**
    ```bash
    mi_array[""]="valor" # No permitido
    ```

### Valores pueden ser:
  * Strings vacíos
  * Números
  * Texto 

### Diferencias con arrays normales

| Característica | Array normal | Asociativo         |
| -------------- | ------------ | ------------------ |
| Índice         | numérico     | string             |
| Orden          | sí           | no garantizado     |
| Declaración    | opcional     | obligatoria (`-A`) |



````python
%%bash
declare -A usuario

usuario["nombre"]="Juan"
usuario["edad"]=23
usuario["ciudad"]="Lima"

printf "Ver valores: %s\n" "${usuario[*]}" # Muestra valores en una sola cadena *

# Recorriendo clave - valor
printf "\nUsando for ... \n"
for clave in "${!usuario[@]}"
do
  printf "Clave %s -> %s\n" "$clave" "${usuario[$clave]}"
done

````

````python
%%bash
declare -A contador

palabras=("hola"  "mundo"  "hola" "hola")

for palabra in "${palabras[@]}"
do
  (( contador["$palabra"]++ ))
done

for clave in "${!contador[@]}"
do
  printf "%s -> %d\n" "$clave" "${contador[$clave]}"
done
````
