# EXIT CODES
Cada vez que ejecutas un comando en la terminal, este devuelve un número al terminar:
  * `0`-> todo salió bien
  * distinto de `0` -> ocurrio algún error

> Se puede verificar con `echo $?` para ver el último exit code del comando anterior ejecutado.

---

## 1. Exit codes comunes
```txt
# 0 → éxito
# 1 → error general
# 2 → uso incorrecto (bad usage)
# 126 → permiso denegado al ejecutar
# 127 → comando no encontrado
# 128+N → terminado por señal N (ej: 130 = Ctrl+C)


exit 0    # salir con éxito
exit 1    # salir con error genérico

```

> Bash es al revés de lo que esperas: en JS/Python, true es `1` y false es `0`. En bash, `0 = éxito = true` y cualquier otro número = error = false. Esto viene de la convención Unix — un proceso exitoso retorna 0, cualquier código de error es distinto de 0.

### Ejemplos

````python
%%bash
ls -l 0-shell.ipynb
echo "Código: $?"

````

````python
%%bash
ls archivo
echo "Código: $?"
````

````python
%%bash
pepepe
echo "Código: $?"
````

## 2. Ejercicios

### Encadenar comandos 

````python
%%bash
echo "Paso 1" && echo "Paso 2"
# Primero es 0 (éxito) por eso se ejecuta el siguiente
````

````python
%%bash
false && echo "Esto no se imprime"
echo $? # No debería imprimirse
````

````python
%%bash
false || echo "Hubo un error"
# El siguiente se ejecuta si falla el primero
````

````python
%%bash
echo "Inicio" && false && echo "Nunca verás esto"
# Primer es 0, segundo es 1 ya no toma en cuenta los siguientes
# Bash evalúa de izquierda a derecha, y cada && depende del anterior
echo $?
````

````python
%%bash
false || echo B || echo C
# || = “ejecuta si falla"
# B se ejecuta
# C no (porque B ya fue éxito)
````

````python
%%bash
true && true && false && echo "D"
echo $? # false cambia todo
````
