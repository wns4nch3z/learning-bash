# PATH
`PATH` es una variable de entorno que le dice al sistema **en qué directorios buscar ejecutables** cuando escribes un comando.

```bash
echo $PATH
```
Salida típica:
```bash
/usr/local/sbin:/usr/local/bin:/usr/sbin:/usr/bin:/sbin:/bin
```
Cuando escribes `nmap`, el sistema busca ese binario en cada uno de esos directorios, en orden.

---

## Clave para pentesting

### Ejecución de herramientas sin ruta completa
Puedes correr:
```bash
nmap
```
en vez de:
```bash
/usr/bin/nmap
```
Revisar ruta de un comando:
```bash
which comando
# o
command -v comando
```

---

### PATH Hijacking
Esto es una técnica clásica de escalada de privilegios.

#### ***Idea***
Si un programa ejecuta otro comando sin ruta absoluta (ej. `ls`, `cat`, `python`), puedes **inyectar un binario malicioso** antes en el `PATH`.

Ejemplo:
```bash
export PATH=/tmp:$PATH
```
Luego creas:
```bash
echo "id" > /tmp/ls
chmod +x /tmp/ls
```
Si un programa privilegiado ejecuta `ls`, podría ejecutar **tu versión maliciosa**.

#### Enumeración
Siempre revisar:
```bash
echo $PATH
```
Busca:
  * Directorios escribibles (`/tmp`, `/home/user`)
  * Directorios raros o personalizados
  * Orden incorrecto (ej: `/home/user/bin` antes que `/usr/bin`)

#### Añadir herramientas propias
Puedes agregar tus scripts:
```bash
export PATH=$PATH:/home/tuusuario/tools
```
Así ejecutas tus herramientas como comandos normales.
