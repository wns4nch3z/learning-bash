# Ejecución
Es especial que el archivo tenga permisos de ejecución porque de nada vale haber agregado el `shebang`.

## Permisos
Revisar: [Permisos](../notebooks/10-permisos.ipynb)

Dar permisos de ejecución:
```bash
chmod +x script.sh
# Revisar permisos 
ls -l 
```

Flujo completo para crear y ejecutar un script:
```bash
# 1 — crear el archivo
nano deploy.sh          # o vim, o cualquier editor

# 2 — dar permiso de ejecución
chmod +x deploy.sh

# 3 — verificar
ls -l deploy.sh
# -rwxr-xr-x  1 tu_usuario  staff  142 mar 17 10:22 deploy.sh
#  ^^^                                                 ↑ el x confirma que es ejecutable

# 4 — ejecutar (el shebang entra en acción)
./deploy.sh
```
Una distinción que ahorra confusión: `Permission denied` al intentar `./script.sh` casi siempre significa que falta el bit `x`. `No such file or directory` en el shebang significa que el intérprete no existe en esa ruta. Son dos errores distintos con causas distintas.
