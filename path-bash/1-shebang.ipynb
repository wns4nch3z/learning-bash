# SHEBANG
El shebang es la primera línea de un script que le dice al sistema qué **programa debe usar para ejecutarlo**.

```bash
#!<ruta-del-interprete>
```
Ejemplo
```bash
#!/bin/bash
```

### ¿Cómo funciona internamente?
Cuando ejecutas un archivo como: `./mi_script`

El kerne de inux hace lo siguiente:
1. Lee la primera línea del archivo
2. Si esa línea comienza con `#!` (shebang), toma lo que sigue como el intérprete
3. Entonces Linux ejecuta internamente `<interprete> <script>`: `/bin/bash mi_script.sh`


### Formas de usar

#### 1. `#!/bin/bash`
  * Bash casi siempre está en `/bin`
  * Funciona en la mayoría de destribuciones linux

#### 2. `#!/usr/bin/env` | PORTABLE
`env` busca el intérprete en el PATH del sistema, en lugar de depender de una ruta fija.

Esto evita problemas cuando el intérprete no está en un lugar estándar.

Funciona con bash, python, node perl, ruby, entre otros.

  * No sabes dónde está instalado bash
  * Trabajas en MacOS o BSD
  * Usas contenedores o entornos personalizados

```bash
# USO 
#!/usr/bin/env python3
#!/usr/bin/env node
#!/usr/bin/env ruby

```
