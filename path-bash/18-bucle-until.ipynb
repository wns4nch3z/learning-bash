# UNTIL
`until` es un bucle que **se ejecuta mientras una condición sea falsa**.

Es básicamente lo contrario de `while`.

  * `while` → se repite mientras la condición sea verdadera
  * `until` → se repite hasta que la condición sea verdadera

--- 

## Sintaxis
```bash
until [ comdición ]
do
  comandos
done
```
Forma compacta:
```bash
until condición; do
  comandos
done
```

***Como funciona:***
  * Se evalúa la condición
  * Si la condición es **falsa**, se ejecuta el bloque.
  * Se repite hasta que la condición sea verdadera

| Tipo  | Se ejecuta cuando...      |
| ----- | ------------------------- |
| while | la condición es VERDADERA |
| until | la condición es FALSA     |

`until` es `while` negado. until condicion = while ! condicion. Brilla en scripts de espera de servicios donde la semántica "esperar hasta que esté listo" es más natural.

## Ejemplos

### 1. Contador
```bash
contador=1

until [ $contador -gt 5 ] 
do
  echo "Número $contador"
  (( contador++ ))
done
```
Explicación:
  * El bucle continúa hasta que `contador > 5`
  * Mientras eso sea falso, imprime los números del 1 al 5  

### 2. Esperar a que exista un archivo
```bash
until [ -f archivo.txt ]
do
  echo "Esperando a que exista archivo.txt..."
  sleep 2
done

echo "¡El archivo ya existe!"
```
Aquí:
  * El bucle se repite que el archivo exista (`-f`)
  * Útil para scripts que dependen de recursos externos


### 3. Intentos hasta éxito
```bash
until ping -c 1 google.com &> /dev/null
do
  echo "Sin conexión... reintentando"
  sleep 3
done

echo "Conexión establecida"
```
