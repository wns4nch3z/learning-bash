# Funciones
 
## ¿Qué es?
Una función es un bloque de código reutilizable que ejecuta una tarea específica.

Sirve para:
  * Evitar repetir código
  * Hacer scripts más ordenados
  * Separar lógica en partes reutilizables
  * Facilitar debugging
  * Crear scripts profesionales

> En scripts reales, todo suele estar dividido en funciones.

## Sintaxis básica
### Forma recomendada
```bash
nombre_funcion() {
    comandos
}
# Usarlo
nombre_funcion
```

### Forma alternativa (menos usada hoy)
```bash
function nombre_funcion {
    comandos
}
```

### Buenas prácticas
  * Usa nombres `descriptivos` (leer_archivos, procesar_datos).  
  * Mantén funciones con `una sola responsabilidad`.
  * Declara funciones antes de `usarlas`.

````python
%%bash
# Ejemplo simple
saludar() {
    echo "Hola, usuario"
}
saludar # Llamando a la función
````

  * Se creó una función: `saludar()`
  * Se ejecutó llamándola: `saludar`

````python
%%bash
mostrar_banner() {
    echo "======================"
    echo "   SCRIPT ANALYZER   "
    echo "======================"   
}

mostrar_banner
````

### Usando parámetros
Bash no declara parámetros los recibe así:

| Variable | Significado      |
| -------- | ---------------- |
| `$1`     | primer argumento |
| `$2`     | segundo          |
| `$@`     | todos            |
| `$#`     | cantidad         |

````python
%%bash
mi_funcion() {
    echo "Hola $1"
}
# Usando con parámetros
mi_funcion "Carla"
````

## Conceptos

### Scope (global y local)
Por defecto las variables son globales.

````python
%%bash
x=10

mi_funcion() {
  echo "$x"
}
mi_funcion
````

`local` limita la variable a la función. Solo se usa en funciones.

````python
%%bash
mi_funcion() {
    local x=5
    echo "Dentro: $x"
}

x=10 # Variable global
mi_funcion # Variable local no afecta al global
echo "Fuera: $x"
````

### Flujo de programa

#### `return` -> controla funciones
En Bash, `return` no devuelve valores como en otros lenguajes (Python, JS, etc.).

Solo devuelve un código de estado numérico:
  * `0` → éxito
  * `1–255` → error u otros estados


````python
%%bash
mi_funcion() {
    return 0
}

mi_funcion
echo $?
````

***¿Qué pasa aquí?***
  * Se ejecuta `mi_function`
  * `return 0` indica que todo salió bien
  * `$?` guarda el `último código de salida`

Resultado: `0`

````python
%%bash
# Cambiando el código de retorno
mi_funcion() {
    return 5
}

mi_funcion
echo "$?"
````

> Bash solo permite valores entre `0-255`

````python
%%bash
# Uso práctico
es_par() {
  if (( "$1" % 2 == 0)); then
    return 0 # Verdadero
  else
    return 1 # Falso
  fi    
}

es_par 4
echo "$?"
es_par 5
echo "$?"
````

````python
%%bash
# Usado correctamente
es_par() {
  if (( "$1" % 2 == 0)); then
    return 0 # Verdadero
  else
    return 1 # Falso
  fi    
}

if es_par 4; then # Si es 0 entra al then
  echo "Es par"
else
  echo "No es par" # Cualquier otro entra aquí
fi
````

#### `exit`
Controla el script completo, `exit` también devuelve código.

````python
%%bash
echo "Inicio"
exit 0
echo "Nunca se ejecuta"
````

  * `exit 0` -> termina **todo el script** con éxito
  * `exit 1` -> termina **todo el script** con error

> Diferencia con `return 0/1` → termina solo la función y devuelve estado

### `export`
Pasar variables a otros procesos.

````python
%%bash
mensaje="Hola"
bash -c 'echo "Mensaje: $mensaje"' # No ve la variable

export mensaje="Hola"     # ahora sí lo ve
bash -c 'echo "Mensaje: $mensaje"'
````

## Combinación
```bash
archivo_existe() {
    local file="$1"

    if [[ -f "$file" ]]; then
        return 0
    else
        return 1
    fi
}

export RUTA="test.txt"

echo "Inicio"

if archivo_existe "$RUTA"; then
    echo "El archivo existe"
else
    echo "No existe, saliendo..."
    exit 1 # Termina el script completo
fi

echo "Fin del script"
```
