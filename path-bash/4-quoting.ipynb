# Expansión y Quoting


## Concepto

### 1. Expansión
El proceso en que Bash toma lo que escribiste y lo transforma en otra cosa antes de ejecutar el comando.

Idea simple:
```bash
echo $nombre
```
Bash lo convierte internamente en:
```bash
echo Juan
```
Eso es espansión.

> Expansión = “Bash reescribe tu comando antes de ejecutarlo”

#### Exp. de variables
```bash
nombre="Carlos"
echo $nombre
```
Bash reemplaza:
```
$nombre -> Carlos
```

#### Exp. de comandos
```bash
echo $(date)
```
Bash ejecuta date primero y reemplaza:
```
$(date) -> resultado del comando
```

#### Exp. de comodines (globbing)
```bash
ls *.txt
```
Bash reemplaza
```
*.txt -> archivo.txt archivo2.txt
```

#### Exp. aritmética
```bash
echo $(( 2 + 3 ))
```
Bash reemplaza:
```
S(( 2+ 3 )) -> 5
```

#### Word splitting (División por espacios)
```bash
nombre="Juan Perez"
echo $nombre
```
Bash hace:
```
"Juan Perez" -> "Juan" "Perez"
```

### 2. QUOTING
En Bash, quoting (o citado) es la forma de decirle al shell:

> “No interpretes esto como código, trátalo como texto literal (o controla cómo lo interpretas)”

Es clave porque Bash interpreta caracteres especiales ($, *, espacios, etc.), y el quoting te permite controlar ese comportamiento.

Idea simple:
  * Sin quoting -> Bash interpreta
  * Con quoting -> tú decides qué se interpreta

#### Tipos de quoting

1. Sin comillas (peligrosos)
    ```bash
    echo $nombre
    ```
    Bash:
      * Expande variables
      * Divide por espacios (word splitting)
      * Expande comodines (`*`)
      

2. Comillas dobles `" "` (más usado)
    ```bash
    echo "$nombre"
    ```
      * Expande variables (`$nombre`)
      * No divide por espacios 
      * No expande `*` como glob
 
3. Comillas simples `' ' ` (modo literal)
     ```bash
     echo '$nombre'
     ```
       * No expande variables
       * No interpreta nada
      Salida:
      ```ouput
      Hola $nombre
      ```
      > Es como decir: “imprime exactamente esto”


### 3. Scaping
#### Escape con `\`
Sirve para escapar un solo carácter

```bash
echo "Precio \$100"
```
```ouput
Precio $100
```

#### `\n`,`\t`,`\r`
Newline `\n`, tab `\t`, retorno `\r` solo dentro de `$'...'` o `echo -e "..."`
```bash
echo -e "Linea1\nLinea2"
```

#### ANSI-C quoting
Interpreta \n \t \\ dentro de simples
```bash
sep=$'\t' # Tab real en una variable
```

#### `\` (newline)
Continúa el comando en la siguiente línea
```bash
curl -X POST \
    -H "Content-Type: json" \
    https://api.ejemplo.com

```
