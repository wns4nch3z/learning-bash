# IFS
`IFS` significa **Internal Field Separator** (Separador Interno de Campos)

> Es una variable del sistema que le dice a Bash cĂłmo dividir texto en partes.

Por defecto, `IFS` incluye:
  * espacio `" "`
  * tabulaciĂłn `\t`
  * salto de lĂ­nea `\n`

---

## SET 
El comando `set` en Bash tiene varios usos, pero el que usaremos aquĂ­ es:

> asignar valores a los **parĂˇmetros posicionales** es decir `$1`, `$2`; `$3`, ect.

### Ejemplo bĂˇsico
```bash
set -- uno dos tres
echo $1
echo $2
echo $3
```
Resultado:
```ouput
uno
dos
tres
```

BĂˇsicamente: `$1 = uno`, `$2 = dos`, `$3 = tres`

#### `--`
El `--` le dice a bash:

> â€ślo que viene ya no son opciones, son valores normalesâ€ť

Evita errores si un valor empieza con `-`.

#### Con variable (entra IFS)
```bash
texto="uno dos tres"
```
Bash: 
  1. Toma la variable
  2. La divide usando `IFS`
  3. Asigna cada parte a `$1`, `$2`, `$3`

## Combinando 
```bash
texto="a,b,c"
IFS="," # Cambiamos la forma en que separa
set -- texto

echo $1
echo $2
echo $3
```
AquĂ­:
  * `IFS` separa
  * `set` guarda en `$1, $2, $3`

Resetear parĂˇmetros: `set --` este limpia todos `$1`, `$2`, etc.

#### Uso prĂˇctico
Procesando lĂ­neas:
```bash
IFS=":"

read usuario clave uid gid resto <<< "root:x:0:0:root:/root:/bin/bash""

echo $usuario
echo $uid
```
> `<<<` pasa como stdin los strings

#### Restaurar IFS (importante)
Si cambia `IFS`, debes devolverlo a su valor original.
```bash
OLD_IFS=$IFS
IFS=","

# tu cĂłdigo

IFS=$OLD_IFS
```
Evita errores en scripts mĂˇs grandes.

#### Caso prĂˇctico
```bash
datos="manzanas|pera|uva"
IFS="|"
set -- $datos

echo "Frutas 1: $1"
echo "Frutas 2: $2"
echo "Frutas 3: $3"

```

  * `IFS` define cĂłmo se va a dividir el texto
  * `set` toma ese texto ya dividido y lo guarda en $1, $2, $3
