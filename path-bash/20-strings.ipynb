# Manipulación de strings


## Formas

### 1. `${#VAR}` -> Longitud del string
Sirve para saber cuántos caracteres tiene una variable.

````python
%%bash
texto="scripting"

echo ${#texto}
````

````python
%%bash
# Ejemplo
password="hunter123"

if [ ${#password} -lt 7 ]; then
  echo "Password muy corto"
else
  echo "Password válido"
fi

````

### 2. `${VAR:pos:len}` -> Extraer substring
Sirve para cortar una parte del string.

Formato:
```bash
$[VAR:posicion:longitud]
```

> Posición siempre empieza en `0`.

````python
%%bash
texto="pentesting"
#pentesting
#0123456789


echo ${texto:4:4}
# toma 4 caracteres
echo ${texto: -3} 
# toma desde forma inversa
````

### 3. `${VAR/old/new}` -> Reemplazar texto
Sirve para **reemplazar la primera coincidencia**.

Formato:
```bash
${VAR/old/new}
```

````python
%%bash
texto="Hola mundo"

echo ${texto/mundo/linux}
````

Reemplazar todas las coincidencias `${VAR//old/new}`

````python
%%bash
texto="aaa bbb aaa"

echo ${texto//aaa/xxx}
````

### 4. `${VAR##*/}` -> Obtener nombre de archivo
Sirve para quitar todo antes del último `/`.

````python
%%bash
ruta="/home/user/documentos/reporte.txt"

echo ${ruta##*/}
````

Es equivalente a `basename` pero más rápido porque no usa **procesos externos**.

```bash
##   → elimina lo más largo posible
*/   → todo hasta el último /
```

---

### 5. `${VAR^}` 
Convierte la primera letra del string en mayúscula.

````python
%%bash
nombre="edgar"

echo ${nombre^}
````

### 6. `${VAR^^}`
Convertir TODO a mayúsculas

````python
%%bash
mensaje="Hola Mundo"

echo ${mensaje^^}
````

### 7. `${VAR,,}`
Convertir a minúsculas

````python
%%bash
texto="H4CK1NG L1NUX"

echo ${texto,,}
````

### 8. `${VAR:-default}`
Valor por defecto

> Si la variable está vacía o no existe -> usa valor por defecto.

````python
%%bash
usaurio=""

echo ${usuario:-invitado}
````

````python
%%bash
usuario="root"

echo ${usuario:-invitado}
````

Ejemplo de scripts
```bash
nombre=${1:-archivo.txt}
echo "Procesando $nombre"

```
Significa:
  * Si el usuario no pasa argumento
  * usar `"archivo.txt"`

 Ejemplo: 
  * `script.sh` -> Procesando archivo.txt
  * `script.sh datos.log` -> Procesando datos.log 

### 9. `${!VAR}`
Variables indirectas

Significa:
> Usar el valor de una variable como nombre de otra variable.

````python
%%bash
var1="mensaje"

mensaje="Hola Linux"

echo ${!var1}

````

> `${!var1}` se convierte en ${mensaje}. Esto es indirección.

````python
%%bash
# Uso real
user1="alice"
user2="bob"
user3="charlie"

for i in 1 2 3
do 
  var="user$i"
  echo ${!var}
done
````

### 10. `${VAR//caracter/}`
Eliminar caracteres específicos.

Ejemplo:
```bash
${VAR//!/}
```
> Sin valor nuevo → elimina.


````python
%%bash
texto="archivo==.txt"

texto="${texto//=/}"
echo $texto
````

Eliminar caracteres peligrosos:
```bash
limpio="${reemplazar//[!@#$%^&*()+=]/}"
```
Eliminar cualquiera de:
```bash
! @ # $ % ^ & * ( ) + =
```
Esto es sanitización real.

### Diferencia de `#`, `%`
La regla es: `#` y `%` apuntan hacia el inicio y el final respectivamente. Simple y `##` o `%%` son la versión greedy.

````python
%%bash
archivo="www.facebook.com"
# Con #
x_corto="${archivo#*.}"
x_largo="${archivo##*.}"

# Con %
y_corto="${archivo%.*}"
y_largo="${archivo%%.*}"

echo "-------------------"
echo ""
echo "Usando #"
echo "Corto: $x_corto"
echo "Largo: $x_largo"
echo ""
echo "Usando %"
echo "Corto: $y_corto"
echo "Largo: $y_largo"
````

> Estos reemplazan `basename` y `dirname` sin crear procesos hijos — importante en bucles sobre miles de archivos.
