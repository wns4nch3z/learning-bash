# CASE
`case` es como un **switch** en otros lenguajes.

Sirve para ejecutar código dependiendo del valor de una variable.

## Sintaxis básica
```bash
case "$variable" in
  valor1)
    comandos
    ;;
  valor2)
    comandos
    ;;
  *)
    comandos_por_defecto 
    ;;
esac
```

> Si escribe otra cosa → entra en `*` (comodín = cualquier cosa)

````python
%%bash
# EJEMPLO BÁSICO
opcion="start"

echo -e "\nOPCIONES\n"
#read -p "Opción: " opción

case $opcion in
  start)
    echo -e "Opción \"$opcion\"\n"
    echo "Iniciando servicio"
    ;;
  stop)
    echo -e "Opción $opcion\n"
    echo "Deteniendo servicio" 
    ;;
  *) 
    echo "Opción no válido"   
    ;;
esac
````

## Otros usos
### Múltiples opciones
```bash
case "$opcion" in
  start|run|go)
    echo "Ejecutando..."
    ;;
  stop|end)
    echo "Deteniendo..."
    ;;
  *)
    echo "No reconocido"
    ;;
esac
```

### Patrones
`case` permite usar patrones tipo glob
```bash
read archivo

case "$archivo" in
  *.txt)
    echo "Es un archivo de texto"
    ;;
  *.jpg|*.png)
    echo "Es una imagen"
    ;;
  *)
    echo "Tipo desconocido"
    ;;
esac
```
> Aquí no comparas texto exacto, sino patrones.

---

### Rango de caracteres
Se puede usar rangos como `[a-z]`, `[0-9]`, etc.
```bash
read letra

case "$letra" in
  [a-z])
    echo "Es una letra minúscula"
    ;;
  [A-Z])
    echo "Es una letra mayúscula"
    ;;
  [0-9])
    echo "Es un número"
    ;;
  *)
    echo "Otro tipo de carácter"
    ;;
esac
```
> Rangos grandes `[1-9]|[1-9][0-9]` representa 1 entre 99

### Coincidencias parciales | comodines
```bash
read palabra

case "$palabra" in
  hola*)
    echo "Empieza con hola"
    ;;
  *mundo)
    echo "Termina con mundo"
    ;;
  *bash*)
    echo "Contiene bash"  
    ;;
esac
```

### Comportamiento de sentencia | fall-through
En una estructura `case`, `fall-through` significa que después de ejecutar el bloque de código de un patrón coincidente, la ejecución no se detiene, sino que continúa con el siguiente bloque.

Por defecto, Bash no permite el `fall-through`; el comando `;;` hace que la sentencia termine después de la primera coincidencia.

Para habilitar el fall-through, se usan terminadores especiales:
  * `;&`: Ejecuta el siguiente bloque **sin importar si su patrón coincide**.
  * `;;&`: Ejecuta el siguiente bloque **solo si su patrón también coincide**.

| Símbolo | Qué hace                          |
| ------- | --------------------------------- |
| `;;`    | termina (tipo `break`)            |
| `;&`    | sigue al siguiente SIN evaluar    |
| `;;&`   | sigue al siguiente PERO evaluando |


````python
%%bash
x="abc"

case $x in
  a*)
    echo "Empieza con a"
    ;;& # Continua pero evalúa
  *c)
    echo "Termina en c"
    ;; # Termina
esac
````

## EJEMPLO REAL

### Analizador de log 
Queremos analiza una línea y detectar:
  * si empieza con `ERROR`
  * si contiene un número
  * si termina en `!`

Una misma línea puede cumplir **varias condiciones**.

````python
%%bash
# EJEMPLO 
linea="ERROR 404!"

case $linea in
  ERROR*)
    echo "Es un error"
    ;;&
  *[0-9]*)
    echo "Contiene números"
    ;;&
  *!)
    echo "Termina en !"
    ;;
esac



````

Paso a paso:

  1. `ERROR*` → ✔ coincide
    imprime: Es un error
  2. `;;&` → sigue pero evaluando el siguiente patrón
  3. `*[0-9]*` → ✔ coincide
    imprime: Contiene números
  4. `;;&` otra vez → evalúa siguiente
  5. `*!` → ✔ coincide
    imprime: Termina en !

Si usarás `;;` terminaría en la primera coincidencia y si usarás `;&` pasa al siguiente sin evaluar dando falsos positivos.
